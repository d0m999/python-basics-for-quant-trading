## 2. 乘法计数法则

### 📐 定义
对于一个事件序列，其中第一个事件可以有 $n_1$ 种发生方式，第二个事件可以有 $n_2$ 种发生方式，第三个事件可以有 $n_3$ 种发生方式，以此类推，事件结果的总数为 $n_1 \times n_2 \times n_3 \times \cdots$。

### 📖 书中的例子
> 超级百万彩票的规则是花费 2 美元从 1~70 个数字中选取 5 个数字，并从 1~25 个数字中选取一个特别数字，在该规则下中头奖的概率为 1/302,575,350。

````python
# ========== 步骤 1: 乘法计数法则 — 密码破解 ==========

# 场景：一个 5 位密码，键盘有 92 个不同字符
n_chars = 92  # 每个位置可选的字符数
n_positions = 5  # 密码位数

# 应用乘法计数法则
total_passwords = n_chars ** n_positions

print(f"\n📊 步骤 1: 乘法计数法则 — 密码破解")
print(f"  每个位置可选字符数 = {n_chars}")
print(f"  密码位数 = {n_positions}")
print(f"  计算公式: {n_chars} × {n_chars} × {n_chars} × {n_chars} × {n_chars}")
print(f"  实际计算: {n_chars}^{n_positions} = {total_passwords:,}")
print(f"\n💡 第一次尝试就破解密码的概率 = 1/{total_passwords:,} ≈ {1/total_passwords:.2e}")
````

````python
# ========== 步骤 2: 超级百万彩票 ==========

# 从 1~70 选 5 个数字（不考虑顺序），从 1~25 选 1 个特别数字
n_main = 70  # 主号码池
k_main = 5   # 选取的主号码数
n_special = 25  # 特别号码池

# 主号码的组合数
main_combinations = special.comb(n_main, k_main, exact=True)
# 总可能性 = 主号码组合 × 特别号码
total_mega = main_combinations * n_special

print(f"\n📊 步骤 2: 超级百万彩票中奖概率")
print(f"  主号码: 从 {n_main} 个中选 {k_main} 个")
print(f"  主号码组合数 C({n_main},{k_main}) = {main_combinations:,}")
print(f"  特别号码: 从 {n_special} 个中选 1 个")
print(f"  总组合数 = {main_combinations:,} × {n_special} = {total_mega:,}")
print(f"\n💡 中头奖概率 = 1/{total_mega:,} ≈ {1/total_mega:.2e}")
print(f"  即约 {total_mega/1e6:.1f} 百万分之一")
````

## 3. 阶乘法则

### 📐 定义
$n$ 个不同元素的排列方式总数为 $n!$（n 的阶乘）：
$$n! = n \times (n-1) \times (n-2) \times \cdots \times 2 \times 1$$

特殊规定：$0! = 1$

### 📖 书中的例子
> "steam" 所包含的字母有 5 个不同的字母，不同排列方式的总数为 $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$。

````python
# ========== 步骤 3: 阶乘法则 ==========

# 计算不同 n 值的阶乘
n_values = list(range(0, 11))
factorials = [math.factorial(n) for n in n_values]

print(f"\n📊 步骤 3: 阶乘计算")
print(f"  {'n':>3} | {'n!':>15}")
print(f"  {'---':>3} | {'---':>15}")
for n, f in zip(n_values, factorials):
    print(f"  {n:>3} | {f:>15,}")

# "steam" 的字母排列
word = "steam"
n_letters = len(word)
permutations = math.factorial(n_letters)
print(f"\n💡 '{word}' 的字母排列方式总数 = {n_letters}! = {permutations}")
print(f"  随机排列恰好按字母顺序 (aemst) 的概率 = 1/{permutations} = {1/permutations:.6f}")
````

````python
# ========== 可视化: 阶乘增长速度 ==========

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(range(len(n_values)), factorials, color='steelblue', edgecolor='white')
ax.set_xticks(range(len(n_values)))
ax.set_xticklabels(n_values)
ax.set_xlabel('n', fontsize=12)
ax.set_ylabel('n! (Factorial)', fontsize=12)
ax.set_title('Factorial Growth: n! for n = 0 to 10', fontsize=14, fontweight='bold')
ax.set_yscale('log')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  阶乘增长非常快！10! = {math.factorial(10):,}")
print(f"  这就是为什么排列问题的结果通常很大")
````

## 4. 排列法则（元素相异）

### 📐 定义
从 $n$ 个元素中无放回地任取 $r$ 个元素排成一列，并考虑它们之间的先后次序，其排列总数为：
$$P(n,r) = \frac{n!}{(n-r)!}$$

### 📖 书中的例子
> 三重彩投注：赛马比赛有 20 匹马参赛，需要以正确顺序选中第 1、2、3 名。
> 排列数 $P(20,3) = \frac{20!}{(20-3)!} = 20 \times 19 \times 18 = 6840$

````python
# ========== 步骤 4: 排列法则 — 三重彩投注 ==========

n_horses = 20  # 参赛马匹数
r_places = 3   # 需要选出的名次数

# 手算排列数
perm_manual = math.factorial(n_horses) // math.factorial(n_horses - r_places)

# 用 scipy 验证
perm_scipy = special.perm(n_horses, r_places, exact=True)

print(f"\n📊 步骤 4: 排列法则 — 三重彩投注")
print(f"  n = {n_horses} 匹马, r = {r_places} 个名次")
print(f"  计算公式: P({n_horses},{r_places}) = {n_horses}! / ({n_horses}-{r_places})!")
print(f"  实际计算: {n_horses} × {n_horses-1} × {n_horses-2} = {n_horses} × {n_horses-1} × {n_horses-2}")
print(f"\n🔬 scipy.special.perm 验证:")
print(f"  手算 P({n_horses},{r_places}) = {perm_manual:,}")
print(f"  scipy P({n_horses},{r_places}) = {perm_scipy:,}")
print(f"\n  ✅ 验证通过！")
print(f"\n💡 随机选 3 匹马中奖的概率 = 1/{perm_scipy:,} ≈ {1/perm_scipy:.6f}")
````

## 5. 排列法则（元素重复）

### 📐 定义
从 $n$ 个元素中无放回地任取 $n$ 个元素排成一列，其中包含重复元素，其排列总数为：
$$\frac{n!}{n_1! \times n_2! \times \cdots \times n_k!}$$

其中，$n_1$ 个元素相同，$n_2$ 个元素相同，…，$n_k$ 个元素相同。

### 📖 书中的例子
> 在一项包含 10 个问题的调查中，有 2 个问题完全相同，还有 3 个问题也完全相同。
> 排列数 = $\frac{10!}{2! \times 3!} = 302,400$

````python
# ========== 步骤 5: 排列法则（重复元素） ==========

n_total = 10  # 总问题数
n_repeat1 = 2  # 第一组重复数
n_repeat2 = 3  # 第二组重复数

# 手算
perm_repeat_manual = math.factorial(n_total) // (math.factorial(n_repeat1) * math.factorial(n_repeat2))

# 验证
perm_repeat_check = math.factorial(n_total) // (math.factorial(n_repeat1) * math.factorial(n_repeat2))

print(f"\n📊 步骤 5: 排列法则（重复元素）")
print(f"  总元素数 n = {n_total}")
print(f"  第一组重复数 = {n_repeat1}")
print(f"  第二组重复数 = {n_repeat2}")
print(f"  计算公式: {n_total}! / ({n_repeat1}! × {n_repeat2}!)")
print(f"  实际计算: {math.factorial(n_total):,} / ({math.factorial(n_repeat1)} × {math.factorial(n_repeat2)})")
print(f"  结果 = {perm_repeat_manual:,}")
print(f"\n💡 10 个问题有 {perm_repeat_manual:,} 种不同的排列")
````

## 6. 组合法则

### 📐 定义
从 $n$ 个元素中无放回地任取 $r$ 个元素，但不考虑它们之间的先后次序，其组合总数为：
$$C(n,r) = \binom{n}{r} = \frac{n!}{(n-r)! \times r!}$$

### 📖 书中的例子
> 从 1~35 中选取 5 个不同的数字赢得彩票头奖。
> 组合数 $C(35,5) = \frac{35!}{(35-5)! \times 5!} = 324,632$

````python
# ========== 步骤 6: 组合法则 — 彩票 ==========

n_pool = 35  # 号码池
k_select = 5  # 选取数

# 手算组合数
comb_manual = math.factorial(n_pool) // (math.factorial(n_pool - k_select) * math.factorial(k_select))

# 用 scipy 验证
comb_scipy = special.comb(n_pool, k_select, exact=True)

print(f"\n📊 步骤 6: 组合法则 — 彩票")
print(f"  从 {n_pool} 个数字中选 {k_select} 个（不考虑顺序）")
print(f"  计算公式: C({n_pool},{k_select}) = {n_pool}! / ({n_pool}-{k_select})! × {k_select}!")
print(f"\n🔬 scipy.special.comb 验证:")
print(f"  手算 C({n_pool},{k_select}) = {comb_manual:,}")
print(f"  scipy C({n_pool},{k_select}) = {comb_scipy:,}")
print(f"\n  ✅ 验证通过！")
print(f"\n💡 中头奖概率 = 1/{comb_scipy:,} ≈ {1/comb_scipy:.6f}")
````

## 7. 排列 vs 组合：如何选择？

### 📐 核心区别
| 特性 | 排列 (Permutation) | 组合 (Combination) |
|------|-------------------|-------------------|
| 顺序 | ✅ 考虑顺序 | ❌ 不考虑顺序 |
| 公式 | $P(n,r) = \frac{n!}{(n-r)!}$ | $C(n,r) = \frac{n!}{(n-r)! \times r!}$ |
| 例子 | 密码、赛马名次 | 委员会、彩票号码 |
| 结果 | 更大 | 更小 |

### 📖 书中的例子
> 谷歌公司任命 3 名官员（CEO、主席、COO）：需要排列（顺序重要）
> 设立 3 人规划委员会：需要组合（顺序不重要）

````python
# ========== 步骤 7: 排列 vs 组合对比 ==========

n_candidates = 8  # 候选人数
r_select = 3  # 选取人数

# 排列：官员任命（顺序重要）
perm_result = special.perm(n_candidates, r_select, exact=True)
# 组合：委员会（顺序不重要）
comb_result = special.comb(n_candidates, r_select, exact=True)

print(f"\n📊 步骤 7: 排列 vs 组合对比")
print(f"  从 {n_candidates} 个候选人中选 {r_select} 人")
print(f"\n  🔹 官员任命（排列，顺序重要）:")
print(f"    P({n_candidates},{r_select}) = {perm_result:,}")
print(f"  🔹 委员会（组合，顺序不重要）:")
print(f"    C({n_candidates},{r_select}) = {comb_result:,}")
print(f"\n  排列 / 组合 = {perm_result}/{comb_result} = {perm_result/comb_result:.0f}")
print(f"  这正好等于 {r_select}! = {math.factorial(r_select)}")
print(f"\n💡 排列总是 ≥ 组合，因为排列考虑了更多的顺序可能性")
````

````python
# ========== 可视化: 排列 vs 组合 ==========

ns = range(2, 15)
perms = [special.perm(n, 3, exact=True) for n in ns]
combs = [special.comb(n, 3, exact=True) for n in ns]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(list(ns), perms, 'o-', color='#e74c3c', linewidth=2, markersize=6, label='P(n,3) - Permutation')
ax.plot(list(ns), combs, 's-', color='#2ecc71', linewidth=2, markersize=6, label='C(n,3) - Combination')
ax.set_xlabel('n (Number of Elements)', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Permutation vs Combination: P(n,3) vs C(n,3)', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  红线（排列）始终在绿线（组合）之上")
print(f"  随着 n 增大，两者的差距越来越大")
````

## 8. 综合应用：彩票概率分析

### 🎯 问题
比较不同彩票游戏的中奖概率，理解计数法则的实际应用。

````python
# ========== 步骤 8: 不同彩票概率对比 ==========

# 定义几种彩票规则
lotteries = {
    'Mega Millions': {'main': (70, 5), 'special': 25},
    'Powerball': {'main': (69, 5), 'special': 26},
    'Lotto 6/49': {'main': (49, 6), 'special': 1},
    'Simple 5/35': {'main': (35, 5), 'special': 1},
}

print(f"\n📊 步骤 8: 不同彩票概率对比")
print(f"  {'彩票名称':<18} | {'主号码组合':>15} | {'总组合数':>18} | {'中奖概率':>12}")
print(f"  {'---':<18} | {'---':>15} | {'---':>18} | {'---':>12}")

for name, rule in lotteries.items():
    n, k = rule['main']
    s = rule['special']
    main_comb = special.comb(n, k, exact=True)
    total = main_comb * s
    prob = 1 / total
    print(f"  {name:<18} | {main_comb:>15,} | {total:>18,} | {prob:>12.2e}")

print(f"\n💡 观察：")
print(f"  1. 号码池越大，中奖概率越低")
print(f"  2. 选取的号码越多，中奖概率越低")
print(f"  3. 有特别号码的游戏，概率会额外降低")
````

````python
# ========== 可视化: 彩票概率对比 ==========

fig, ax = plt.subplots(figsize=(10, 6))

names = list(lotteries.keys())
probs = []
for name, rule in lotteries.items():
    n, k = rule['main']
    s = rule['special']
    main_comb = special.comb(n, k, exact=True)
    total = main_comb * s
    probs.append(1 / total)

colors = ['#e74c3c', '#e67e22', '#3498db', '#2ecc71']
bars = ax.bar(names, probs, color=colors, edgecolor='white')
ax.set_ylabel('Probability of Winning', fontsize=12)
ax.set_title('Lottery Winning Probabilities Comparison', fontsize=14, fontweight='bold')
ax.set_yscale('log')
ax.grid(axis='y', alpha=0.3)

# 在柱子上标注数值
for bar, prob in zip(bars, probs):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
            f'{prob:.1e}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  使用对数坐标，可以更清楚地看到概率的差异")
print(f"  Simple 5/35 的中奖概率比 Mega Millions 高约 1000 倍")
````

## 📌 核心概念回顾

### 📌 乘法计数法则 (Multiplication Rule)
- **定义**: 事件序列的总数 = 各阶段选择数的乘积
- **公式**: $n_1 \times n_2 \times \cdots \times n_k$
- **含义**: 适用于独立的多阶段决策

### 📌 阶乘 (Factorial)
- **定义**: $n! = n \times (n-1) \times \cdots \times 1$
- **公式**: $0! = 1$，$n! = n \times (n-1)!$
- **含义**: n 个不同元素的全排列数

### 📌 排列 (Permutation)
- **定义**: 从 n 个元素取 r 个，考虑顺序
- **公式**: $P(n,r) = \frac{n!}{(n-r)!}$
- **含义**: 顺序不同的结果视为不同

### 📌 组合 (Combination)
- **定义**: 从 n 个元素取 r 个，不考虑顺序
- **公式**: $C(n,r) = \frac{n!}{(n-r)! \times r!}$
- **含义**: 顺序不同的结果视为相同

### 🔗 完整流程
```
问题识别 → 选择法则 → 代入公式 → 验证结果
    ↓           ↓          ↓          ↓
  有顺序吗？ 排列/组合    阶乘计算   scipy验证
```

### 📝 检验指标汇总

| 法则 | 公式 | 适用场景 | 顺序要求 |
|------|------|----------|----------|
| 乘法法则 | $n_1 \times n_2$ | 多阶段事件 | 每阶段独立 |
| 阶乘 | $n!$ | 全排列 | 考虑顺序 |
| 排列 | $P(n,r)$ | 取 r 个元素 | 考虑顺序 |
| 组合 | $C(n,r)$ | 取 r 个元素 | 不考虑顺序 |

## ❌ 常见误区

### ❌ 误区 1: 混淆排列和组合
**✓ 正确理解**: 关键看"顺序是否重要"。选密码是排列（123 ≠ 321），选委员会是组合（ABC = CBA）。

### ❌ 误区 2: 忘记处理重复元素
当元素有重复时，必须除以重复元素的阶乘，否则会多算。例如 "AABB" 的排列数是 $\frac{4!}{2! \times 2!} = 6$，而不是 $4! = 24$。

### ❌ 误区 3: 误用组合公式计算彩票概率
彩票中奖概率的计算需要考虑是否有"特别号码"。如果彩票有特别号码，总组合数 = 主号码组合数 × 特别号码数。

### ❌ 误区 4: 认为排列结果总是比组合大很多
**✓ 正确理解**: $P(n,r) = C(n,r) \times r!$。当 r 很小时（如 r=1），排列和组合的结果相同。

### ❌ 误区 5: 忽略 0! = 1 的规定
这个规定不是随意的，它保证了公式的一致性。例如 $C(n,0) = \frac{n!}{n! \times 0!} = 1$，表示"什么都不选"只有 1 种方式。
