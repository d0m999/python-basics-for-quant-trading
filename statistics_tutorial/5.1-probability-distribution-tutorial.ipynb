## 1. 场景设定：概率分布

### 🎯 问题
考虑两个新生儿中女婴的数量：
- 可能的结果：0 个女婴、1 个女婴、2 个女婴
- 每个结果都有特定的概率

这就是一个**概率分布**——它描述了随机变量每个可能值的概率。

### 📖 书中的定义
> 随机变量是一个变量（通常用 x 表示），它对于某个过程的每个结果都有一个随机确定的单个值。
> 概率分布是一种描述，它对随机变量的每一个值给出特定的概率。

## 2. 概率分布的基本概念

### 📐 概率分布的三个条件
一个有效的概率分布必须满足：
1. 随机变量 $x$ 是数值型数据，其值是对应的概率
2. $\sum P(x) = 1$（所有概率之和必须是 1）
3. $0 \leq P(x) \leq 1$（每个概率值必须在 0 和 1 之间）

### 📖 书中的例子
> 假设男婴和女婴的出生率相同，考虑两个新生儿中女婴的数量 $x$。
> $x$ 可以是 0、1 或 2，对应的概率分别为 0.25、0.50、0.25。

````python
# ========== 步骤 1: 构建概率分布表 ==========

# "两个新生儿"中女婴数量的概率分布
x_values = np.array([0, 1, 2])  # 随机变量的可能值
p_values = np.array([0.25, 0.50, 0.25])  # 对应的概率

# 创建 DataFrame
df = pd.DataFrame({'x (女婴数量)': x_values, 'P(x)': p_values})

print(f"\n📊 步骤 1: 概率分布表")
print(df.to_string(index=False))

# 验证三个条件
print(f"\n🔬 验证概率分布的三个条件:")
print(f"  条件1: x 是数值型数据 ✓")
print(f"  条件2: ΣP(x) = {p_values.sum():.2f} {'✓' if np.isclose(p_values.sum(), 1.0) else '✗'}")
print(f"  条件3: 所有 P(x) 在 [0,1] 之间 ✓")
print(f"    0 ≤ {p_values.min():.2f} ≤ 1, 0 ≤ {p_values.max():.2f} ≤ 1")
````

````python
# ========== 步骤 2: 概率直方图 ==========

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(x_values, p_values, width=0.6, color='steelblue', edgecolor='white', alpha=0.8)

# 在柱子上标注概率值
for bar, p in zip(bars, p_values):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01,
            f'{p:.2f}', ha='center', va='bottom', fontsize=12, fontweight='bold')

ax.set_xlabel('x (Number of Girls)', fontsize=12)
ax.set_ylabel('P(x) (Probability)', fontsize=12)
ax.set_title('Probability Distribution: Number of Girls in Two Births', fontsize=14, fontweight='bold')
ax.set_xticks(x_values)
ax.set_ylim(0, 0.65)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  每个矩形的面积 = 概率值（因为宽度为 1）")
print(f"  所有矩形面积之和 = 1")
print(f"  最可能的结果是 1 个女婴（概率 0.50）")
````

## 3. 概率分布的均值（期望值）

### 📐 计算公式
离散概率分布的均值（期望值）：
$$\mu = E(x) = \sum [x \cdot P(x)]$$

其中：
- $x$ = 随机变量的每个可能值
- $P(x)$ = 对应的概率
- $\mu$ = 总体均值（期望值）

### 📖 书中的例子
> 两个新生儿中女婴数量的均值 = $0 \times 0.25 + 1 \times 0.50 + 2 \times 0.25 = 1.0$

````python
# ========== 步骤 3: 计算均值（期望值） ==========

# 手算均值
mean_manual = np.sum(x_values * p_values)

# 详细计算过程
print(f"\n📊 步骤 3: 计算均值（期望值）")
print(f"  计算公式: μ = Σ[x × P(x)]")
print(f"\n  计算过程:")
for x, p in zip(x_values, p_values):
    print(f"    {x} × {p:.2f} = {x*p:.2f}")
print(f"  {'='*30}")
print(f"  μ = {' + '.join([f'{x*p:.2f}' for x, p in zip(x_values, p_values)])}")
print(f"  μ = {mean_manual:.2f}")
print(f"\n💡 均值为 {mean_manual:.2f} 的含义：")
print(f"  如果重复无数次'两个新生儿'的实验")
print(f"  平均每次会有 {mean_manual:.2f} 个女婴")
````

## 4. 概率分布的方差和标准差

### 📐 计算公式
**方差公式（易于理解）**：
$$\sigma^2 = \sum [(x - \mu)^2 \cdot P(x)]$$

**方差公式（手算方便）**：
$$\sigma^2 = \sum [x^2 \cdot P(x)] - \mu^2$$

**标准差**：
$$\sigma = \sqrt{\sigma^2}$$

### 📖 书中的例子
> 两个新生儿中女婴数量的方差 = 0.5，标准差 = 0.7

````python
# ========== 步骤 4: 计算方差和标准差 ==========

# 方法1：σ² = Σ[(x - μ)² × P(x)]
variance_method1 = np.sum((x_values - mean_manual)**2 * p_values)

# 方法2：σ² = Σ[x² × P(x)] - μ²
variance_method2 = np.sum(x_values**2 * p_values) - mean_manual**2

# 标准差
std_manual = np.sqrt(variance_method1)

print(f"\n📊 步骤 4: 计算方差和标准差")
print(f"\n  方法1: σ² = Σ[(x - μ)² × P(x)]")
for x, p in zip(x_values, p_values):
    diff_sq = (x - mean_manual)**2
    print(f"    ({x} - {mean_manual:.1f})² × {p:.2f} = {diff_sq:.2f} × {p:.2f} = {diff_sq*p:.4f}")
print(f"  σ² = {variance_method1:.4f}")

print(f"\n  方法2: σ² = Σ[x² × P(x)] - μ²")
for x, p in zip(x_values, p_values):
    print(f"    {x}² × {p:.2f} = {x**2} × {p:.2f} = {x**2*p:.4f}")
print(f"  Σ[x² × P(x)] = {np.sum(x_values**2 * p_values):.4f}")
print(f"  μ² = {mean_manual**2:.4f}")
print(f"  σ² = {np.sum(x_values**2 * p_values):.4f} - {mean_manual**2:.4f} = {variance_method2:.4f}")

print(f"\n🔬 两种方法验证:")
print(f"  方法1 方差 = {variance_method1:.4f}")
print(f"  方法2 方差 = {variance_method2:.4f}")
print(f"  ✅ 验证通过！")

print(f"\n📊 标准差: σ = √{variance_method1:.4f} = {std_manual:.4f}")
````

## 5. 期望值的应用：赌博决策

### 📐 期望值的含义
期望值是无穷次试验的理论平均结果。在赌博中，期望值告诉你每次下注平均会赢/输多少。

### 📖 书中的例子
> 轮盘赌：押 5 美元在 7 号上，期望值 = -0.26 美元
> 花旗骰：押 5 美元在"过关线"上，期望值 = -0.07 美元

````python
# ========== 步骤 5: 期望值应用 — 赌博决策 ==========

# 轮盘赌：押 5 美元在 7 号上
bet = 5
# 输的概率
p_lose_roulette = 37/38
x_lose_roulette = -bet
# 赢的概率（净收益 = 180 - 5 = 175）
p_win_roulette = 1/38
x_win_roulette = 175

# 期望值
ev_roulette = x_lose_roulette * p_lose_roulette + x_win_roulette * p_win_roulette

# 花旗骰：押 5 美元在'过关线'上
p_lose_craps = 251/495
x_lose_craps = -bet
p_win_craps = 244/495
x_win_craps = 5

ev_craps = x_lose_craps * p_lose_craps + x_win_craps * p_win_craps

print(f"\n📊 步骤 5: 期望值应用 — 赌博决策")
print(f"\n  🔹 轮盘赌（押 5 美元在 7 号上）:")
print(f"    输: {x_lose_roulette} 美元 × {p_lose_roulette:.4f} = {x_lose_roulette*p_lose_roulette:.4f}")
print(f"    赢: {x_win_roulette} 美元 × {p_win_roulette:.4f} = {x_win_roulette*p_win_roulette:.4f}")
print(f"    期望值 = {ev_roulette:.2f} 美元")

print(f"\n  🔹 花旗骰（押 5 美元在'过关线'上）:")
print(f"    输: {x_lose_craps} 美元 × {p_lose_craps:.4f} = {x_lose_craps*p_lose_craps:.4f}")
print(f"    赢: {x_win_craps} 美元 × {p_win_craps:.4f} = {x_win_craps*p_win_craps:.4f}")
print(f"    期望值 = {ev_craps:.2f} 美元")

print(f"\n💡 结论：")
print(f"  轮盘赌每次平均输 {abs(ev_roulette):.2f} 美元")
print(f"  花旗骰每次平均输 {abs(ev_craps):.2f} 美元")
print(f"  花旗骰的期望值更高（输得更少），长期来看更划算")
````

## 6. 显著值的判断

### 📐 范围经验法则
- **显著低**：值 $\leq \mu - 2\sigma$
- **显著高**：值 $\geq \mu + 2\sigma$
- **非显著**：值在 $(\mu - 2\sigma)$ 和 $(\mu + 2\sigma)$ 之间

### 📖 书中的例子
> 两个新生儿中女婴数量：$\mu = 1.0$，$\sigma = 0.7$
> 显著高：$\geq 1.0 + 2(0.7) = 2.4$
> 所以两个女婴（=2）不是显著高的值

````python
# ========== 步骤 6: 判断显著值 ==========

# 使用前面计算的均值和标准差
mu = mean_manual
sigma = std_manual

# 显著值范围
low_threshold = mu - 2 * sigma
high_threshold = mu + 2 * sigma

print(f"\n📊 步骤 6: 判断显著值")
print(f"  均值 μ = {mu:.4f}")
print(f"  标准差 σ = {sigma:.4f}")
print(f"\n  显著低: ≤ {low_threshold:.4f}")
print(f"  显著高: ≥ {high_threshold:.4f}")
print(f"  非显著: {low_threshold:.4f} < x < {high_threshold:.4f}")

# 检查每个可能的值
print(f"\n  各值的显著性判断:")
for x in x_values:
    if x <= low_threshold:
        status = "显著低 ❌"
    elif x >= high_threshold:
        status = "显著高 ⬆️"
    else:
        status = "非显著 ✓"
    print(f"    x = {x}: {status}")

print(f"\n💡 结论：")
print(f"  两个女婴（x=2）不是显著高值，因为 2 < {high_threshold:.2f}")
````

````python
# ========== 可视化: 显著值判断 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制概率分布
bars = ax.bar(x_values, p_values, width=0.6, color='steelblue', edgecolor='white', alpha=0.7)

# 标记均值
ax.axvline(x=mu, color='green', linestyle='-', linewidth=2, label=f'Mean μ = {mu:.2f}')

# 标记显著值范围
ax.axvline(x=low_threshold, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'μ - 2σ = {low_threshold:.2f}')
ax.axvline(x=high_threshold, color='red', linestyle='--', linewidth=2, alpha=0.7, label=f'μ + 2σ = {high_threshold:.2f}')

# 填充非显著区域
ax.axvspan(low_threshold, high_threshold, alpha=0.1, color='green', label='Non-significant zone')

ax.set_xlabel('x (Number of Girls)', fontsize=12)
ax.set_ylabel('P(x) (Probability)', fontsize=12)
ax.set_title('Significance Range: Using Range Rule of Thumb', fontsize=14, fontweight='bold')
ax.set_xticks(x_values)
ax.legend(fontsize=10)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  绿线 = 均值 ({mu:.2f})")
print(f"  红色虚线 = 显著值边界")
print(f"  绿色阴影区域 = 非显著值范围")
print(f"  所有可能的值（0, 1, 2）都在非显著区域内")
````

## 7. 扩展：抛硬币实验的概率分布

### 🎯 问题
抛掷 3 枚硬币，记录正面朝上的次数。构建其概率分布。

````python
# ========== 步骤 7: 扩展 — 抛硬币实验 ==========

# 抛 3 枚硬币，正面朝上的次数
n_coins = 3
x_coins = np.arange(0, n_coins + 1)  # 0, 1, 2, 3

# 使用二项分布计算概率
p_coins = stats.binom.pmf(x_coins, n_coins, 0.5)

# 创建概率分布表
df_coins = pd.DataFrame({'x (正面数)': x_coins, 'P(x)': p_coins})

print(f"\n📊 步骤 7: 抛 3 枚硬币的概率分布")
print(df_coins.to_string(index=False))

# 计算均值、方差、标准差
mean_coins = np.sum(x_coins * p_coins)
var_coins = np.sum((x_coins - mean_coins)**2 * p_coins)
std_coins = np.sqrt(var_coins)

print(f"\n📊 统计量:")
print(f"  均值 μ = {mean_coins:.4f}")
print(f"  方差 σ² = {var_coins:.4f}")
print(f"  标准差 σ = {std_coins:.4f}")

# 显著值
print(f"\n📊 显著值判断:")
print(f"  显著低: ≤ {mean_coins - 2*std_coins:.4f}")
print(f"  显著高: ≥ {mean_coins + 2*std_coins:.4f}")
````

````python
# ========== 可视化: 抛硬币概率分布 ==========

fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(x_coins, p_coins, width=0.6, color='steelblue', edgecolor='white', alpha=0.8)

# 在柱子上标注概率
for bar, p in zip(bars, p_coins):
    ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.01,
            f'{p:.3f}', ha='center', va='bottom', fontsize=11, fontweight='bold')

# 标记均值
ax.axvline(x=mean_coins, color='green', linestyle='-', linewidth=2, label=f'Mean = {mean_coins:.2f}')

ax.set_xlabel('x (Number of Heads)', fontsize=12)
ax.set_ylabel('P(x) (Probability)', fontsize=12)
ax.set_title('Probability Distribution: Heads in 3 Coin Flips', fontsize=14, fontweight='bold')
ax.set_xticks(x_coins)
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  最可能的结果是 1 或 2 个正面（概率各 0.375）")
print(f"  全部正面或全部反面的概率最低（各 0.125）")
print(f"  绿线标记了均值位置 ({mean_coins:.2f})")
````

## 📌 核心概念回顾

### 📌 随机变量 (Random Variable)
- **定义**: 对某个过程的每个结果都有一个随机确定的单个值
- **类型**: 离散型（有限或可数个值）和连续型（无穷多个值）
- **含义**: 用大写字母 X 表示，小写 x 表示具体值

### 📌 概率分布 (Probability Distribution)
- **定义**: 描述随机变量每个可能值的概率
- **公式**: 必须满足 $\sum P(x) = 1$ 且 $0 \leq P(x) \leq 1$
- **含义**: 可以用表格、公式或图形表示

### 📌 期望值 (Expected Value)
- **定义**: 无穷次试验的理论平均结果
- **公式**: $\mu = E(x) = \sum [x \cdot P(x)]$
- **含义**: 长期平均结果，不一定是整数

### 📌 方差和标准差
- **方差**: $\sigma^2 = \sum [(x - \mu)^2 \cdot P(x)]$
- **标准差**: $\sigma = \sqrt{\sigma^2}$
- **含义**: 衡量随机变量的离散程度

### 🔗 完整流程
```
定义随机变量 → 构建概率分布 → 计算均值 → 计算方差 → 判断显著值
    ↓              ↓            ↓          ↓          ↓
  确定x的值      验证条件      Σ[x·P(x)]  Σ[(x-μ)²·P(x)]  μ±2σ
```

### 📝 检验指标汇总

| 指标 | 公式 | 含义 |
|------|------|------|
| 均值 μ | $\sum [x \cdot P(x)]$ | 分布的中心位置 |
| 方差 σ² | $\sum [(x - \mu)^2 \cdot P(x)]$ | 分布的离散程度 |
| 标准差 σ | $\sqrt{\sigma^2}$ | 与原始数据同量纲 |
| 显著低 | $\leq \mu - 2\sigma$ | 异常小的值 |
| 显著高 | $\geq \mu + 2\sigma$ | 异常大的值 |

## ❌ 常见误区

### ❌ 误区 1: 混淆样本统计量和总体参数
**✓ 正确理解**: 概率分布描述的是总体，所以均值、方差、标准差是参数（μ, σ², σ），不是统计量（x̄, s², s）。

### ❌ 误区 2: 认为期望值一定是整数
**✓ 正确理解**: 期望值是理论平均值，不一定是实际可能出现的值。例如，5 个新生儿中女婴的期望值是 2.5，但不可能有 2.5 个女婴。

### ❌ 误区 3: 忘记验证概率分布的三个条件
在计算均值和方差之前，必须先验证：(1) x 是数值型，(2) 所有概率之和为 1，(3) 每个概率在 0 和 1 之间。不满足条件的不能称为概率分布。

### ❌ 误区 4: 用范围经验法则判断所有数据
**✓ 正确理解**: 范围经验法则假设数据近似正态分布。对于严重偏态的分布，这个法则可能不适用。

### ❌ 误区 5: 混淆方差的两个公式
**✓ 正确理解**: $\sigma^2 = \sum [(x - \mu)^2 \cdot P(x)]$ 更直观，$\sigma^2 = \sum [x^2 \cdot P(x)] - \mu^2$ 手算更方便。两个公式结果相同。
