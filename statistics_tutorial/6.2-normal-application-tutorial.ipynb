## 1. 场景设定：非标准正态分布

### 🎯 问题
上一节讨论了标准正态分布（$\mu=0, \sigma=1$），但在实际中很难找到这样的数据。

实际问题中的正态分布通常有不同的均值和标准差：
- 男性身高：$\mu = 68.6$ 英寸，$\sigma = 2.8$ 英寸
- 女性身高：$\mu = 63.7$ 英寸，$\sigma = 2.9$ 英寸
- IQ 分数：$\mu = 100$，$\sigma = 15$

### 📖 书中的定义
> 公式 6-2：$z = \frac{x - \mu}{\sigma}$（结果保留 2 位小数）
>
> 任何正态分布中以某个 $x$ 值为边界的面积，与标准正态分布中以相应的 z 分数为边界的面积相同。

### 💡 关键洞察
**标准化**是将任意正态分布转换为标准正态分布的桥梁。一旦标准化，就可以用上一节的方法计算概率。

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 设置中文字体和样式
import matplotlib.font_manager as fm
_cn_fonts = [f.name for f in fm.fontManager.ttflist if any(kw in f.name for kw in ['Hei', 'Song', 'PingFang', 'Arial Unicode', 'Noto Sans CJK', 'SimHei', 'Microsoft YaHei'])]
plt.rcParams['font.sans-serif'] = _cn_fonts[:3] + ['DejaVu Sans'] if _cn_fonts else ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style('whitegrid')

np.random.seed(42)
print("✅ 库导入完成")
````

## 2. 标准化公式

### 📐 标准化公式
将任意正态分布的 $x$ 值转换为标准正态分布的 z 分数：
$$z = \frac{x - \mu}{\sigma}$$

其中：
- $x$ = 原始数据值
- $\mu$ = 总体均值
- $\sigma$ = 总体标准差
- $z$ = 标准化后的 z 分数

### 📐 反向公式
从 z 分数还原为原始值：
$$x = \mu + z \times \sigma$$

### 💡 标准化的意义
标准化后，不同量纲的数据可以在同一尺度上比较。

````python
# ========== 步骤 1: 标准化计算 ==========

# 场景：男性身高服从 N(68.6, 2.8)
mu_male = 68.6  # 均值（英寸）
sigma_male = 2.8  # 标准差（英寸）
x_height = 72  # 我们想知道身高高于 72 英寸的男性比例

# 标准化
z_height = (x_height - mu_male) / sigma_male

print(f"\n📊 步骤 1: 标准化计算")
print(f"  男性身高分布: N(μ={mu_male}, σ={sigma_male})")
print(f"  原始值: x = {x_height} 英寸")
print(f"\n  标准化公式: z = (x - μ) / σ")
print(f"  计算过程: z = ({x_height} - {mu_male}) / {sigma_male}")
print(f"  z = {z_height:.4f}")
print(f"\n💡 身高 {x_height} 英寸对应 z = {z_height:.2f}，即高于均值 {z_height:.2f} 个标准差。")
````

## 3. 求高于某值的比例

### 📐 计算步骤
1. 画正态曲线，标记均值和 $x$ 值
2. 将 $x$ 值标准化为 z 分数
3. 用 scipy 计算概率

### 📖 书中的例子
> 男性身高服从 $N(68.6, 2.8)$。求身高高于 72 英寸的男性比例。

````python
# ========== 步骤 2: 求高于 72 英寸的男性比例 ==========

# 方法1: 用 scipy 直接计算（推荐）
prob_above_72 = 1 - stats.norm.cdf(x_height, mu_male, sigma_male)

# 方法2: 先标准化，再查标准正态分布
prob_above_72_z = 1 - stats.norm.cdf(z_height, 0, 1)

print(f"\n📊 步骤 2: 求身高高于 {x_height} 英寸的男性比例")
print(f"\n  方法1: 直接用 scipy 计算")
print(f"  P(X > {x_height}) = 1 - P(X < {x_height})")
print(f"  P(X > {x_height}) = 1 - {stats.norm.cdf(x_height, mu_male, sigma_male):.4f}")
print(f"  P(X > {x_height}) = {prob_above_72:.4f}")
print(f"  即约 {prob_above_72*100:.2f}% 的男性身高高于 {x_height} 英寸")

print(f"\n  方法2: 先标准化，再查标准正态分布")
print(f"  z = {z_height:.4f}")
print(f"  P(Z > {z_height:.4f}) = {prob_above_72_z:.4f}")
print(f"\n  ✅ 两种方法结果一致！")
````

````python
# ========== 可视化: 男性身高分布 ==========

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 左图: 原始分布
x_range = np.linspace(mu_male - 4*sigma_male, mu_male + 4*sigma_male, 1000)
y_range = stats.norm.pdf(x_range, mu_male, sigma_male)

axes[0].plot(x_range, y_range, 'b-', linewidth=2)
x_fill = np.linspace(x_height, mu_male + 4*sigma_male, 500)
y_fill = stats.norm.pdf(x_fill, mu_male, sigma_male)
axes[0].fill_between(x_fill, 0, y_fill, alpha=0.3, color='#e74c3c',
                     label=f'P(X > {x_height}) = {prob_above_72:.4f}')
axes[0].axvline(x=mu_male, color='steelblue', linestyle='--', alpha=0.7, label=f'Mean = {mu_male}')
axes[0].axvline(x=x_height, color='#e74c3c', linestyle='--', alpha=0.7)
axes[0].set_xlabel('Height (inches)', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)
axes[0].set_title(f'Male Height Distribution N({mu_male}, {sigma_male})', fontsize=12, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(alpha=0.3)

# 右图: 标准正态分布
z_range = np.linspace(-4, 4, 1000)
y_z = stats.norm.pdf(z_range, 0, 1)

axes[1].plot(z_range, y_z, 'b-', linewidth=2)
z_fill = np.linspace(z_height, 4, 500)
y_z_fill = stats.norm.pdf(z_fill, 0, 1)
axes[1].fill_between(z_fill, 0, y_z_fill, alpha=0.3, color='#e74c3c',
                     label=f'P(Z > {z_height:.2f}) = {prob_above_72_z:.4f}')
axes[1].axvline(x=0, color='steelblue', linestyle='--', alpha=0.7, label='Mean = 0')
axes[1].axvline(x=z_height, color='#e74c3c', linestyle='--', alpha=0.7)
axes[1].set_xlabel('z Score', fontsize=12)
axes[1].set_ylabel('Density', fontsize=12)
axes[1].set_title('Standard Normal Distribution N(0, 1)', fontsize=12, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  左图：原始男性身高分布 N({mu_male}, {sigma_male})。")
print(f"  右图：标准化后的标准正态分布 N(0, 1)。")
print(f"  红色区域：身高高于 {x_height} 英寸的比例 = {prob_above_72:.4f}。")
print(f"  两个红色区域的面积相同，这就是标准化的威力。")
````

## 4. 求两个值之间的比例

### 📐 计算公式
$$P(x_1 < X < x_2) = P(Z < z_2) - P(Z < z_1)$$

### 📖 书中的例子
> 美国空军要求飞行员身高在 64 英寸和 77 英寸之间。
> 女性身高服从 $N(63.7, 2.9)$。满足这一身高要求的女性比例是多少？

````python
# ========== 步骤 3: 求两个值之间的比例 ==========

# 女性身高分布
mu_female = 63.7
sigma_female = 2.9

# 身高要求
x_low = 64
x_high = 77

# 标准化
z_low = (x_low - mu_female) / sigma_female
z_high = (x_high - mu_female) / sigma_female

# 计算概率
prob_below_low = stats.norm.cdf(x_low, mu_female, sigma_female)
prob_below_high = stats.norm.cdf(x_high, mu_female, sigma_female)
prob_between = prob_below_high - prob_below_low

print(f"\n📊 步骤 3: 求身高在 {x_low} 和 {x_high} 英寸之间的女性比例")
print(f"  女性身高分布: N(μ={mu_female}, σ={sigma_female})")
print(f"\n  标准化:")
print(f"    对于 x = {x_low}: z = ({x_low} - {mu_female}) / {sigma_female} = {z_low:.4f}")
print(f"    对于 x = {x_high}: z = ({x_high} - {mu_female}) / {sigma_female} = {z_high:.4f}")
print(f"\n  概率计算:")
print(f"    P(X < {x_low}) = {prob_below_low:.4f}")
print(f"    P(X < {x_high}) = {prob_below_high:.4f}")
print(f"\n    P({x_low} < X < {x_high}) = P(X < {x_high}) - P(X < {x_low})")
print(f"    P({x_low} < X < {x_high}) = {prob_below_high:.4f} - {prob_below_low:.4f}")
print(f"    P({x_low} < X < {x_high}) = {prob_between:.4f}")
print(f"\n  即约 {prob_between*100:.2f}% 的女性满足飞行员身高要求。")
print(f"  约 {(1-prob_between)*100:.2f}% 的女性不满足要求。")
````

````python
# ========== 可视化: 女性身高与飞行员要求 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制女性身高分布
x_range = np.linspace(mu_female - 4*sigma_female, mu_female + 4*sigma_female, 1000)
y_range = stats.norm.pdf(x_range, mu_female, sigma_female)

ax.plot(x_range, y_range, 'b-', linewidth=2)

# 填充满足要求的区域
x_fill_between = np.linspace(x_low, x_high, 500)
y_fill_between = stats.norm.pdf(x_fill_between, mu_female, sigma_female)
ax.fill_between(x_fill_between, 0, y_fill_between, alpha=0.3, color='#2ecc71',
                label=f'Meets requirement: {prob_between:.4f}')

# 填充不满足要求的区域
x_fill_below = np.linspace(mu_female - 4*sigma_female, x_low, 500)
y_fill_below = stats.norm.pdf(x_fill_below, mu_female, sigma_female)
ax.fill_between(x_fill_below, 0, y_fill_below, alpha=0.3, color='#e74c3c')

x_fill_above = np.linspace(x_high, mu_female + 4*sigma_female, 500)
y_fill_above = stats.norm.pdf(x_fill_above, mu_female, sigma_female)
ax.fill_between(x_fill_above, 0, y_fill_above, alpha=0.3, color='#e74c3c',
                label=f'Does not meet: {1-prob_between:.4f}')

# 标记边界
ax.axvline(x=x_low, color='#e67e22', linestyle='--', linewidth=2, label=f'Lower limit = {x_low}"')
ax.axvline(x=x_high, color='#e67e22', linestyle='--', linewidth=2, label=f'Upper limit = {x_high}"')
ax.axvline(x=mu_female, color='steelblue', linestyle='--', alpha=0.7, label=f'Mean = {mu_female}"')

ax.set_xlabel('Height (inches)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('Female Height Distribution and Pilot Requirement', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  绿色区域：身高在 {x_low}-{x_high} 英寸之间的女性（满足要求）。")
print(f"  红色区域：身高不在要求范围内的女性。")
print(f"  橙色虚线：身高要求的上下限。")
print(f"  约 {prob_between*100:.1f}% 的女性满足飞行员身高要求。")
````

## 5. 求满足特定概率的临界值

### 📐 计算公式
给定概率，求对应的 $x$ 值：
$$x = \mu + z \times \sigma$$

其中 $z$ 是给定概率对应的 z 分数。

### 📖 书中的例子
> 男性身高服从 $N(68.6, 2.8)$。求第 95 百分位数对应的身高。

````python
# ========== 步骤 4: 求第 95 百分位数 ==========

# 给定概率
prob_95 = 0.95

# 方法1: 用 scipy 直接计算
x_95 = stats.norm.ppf(prob_95, mu_male, sigma_male)

# 方法2: 先求 z 分数，再还原
z_95 = stats.norm.ppf(prob_95, 0, 1)
x_95_manual = mu_male + z_95 * sigma_male

print(f"\n📊 步骤 4: 求第 95 百分位数对应的身高")
print(f"  男性身高分布: N(μ={mu_male}, σ={sigma_male})")
print(f"  给定概率: P(X < x) = {prob_95}")
print(f"\n  方法1: 直接用 scipy 计算")
print(f"  x = {x_95:.4f} 英寸")
print(f"\n  方法2: 先求 z 分数，再还原")
print(f"  z_{{{prob_95}}} = {z_95:.4f}")
print(f"  x = μ + z × σ = {mu_male} + {z_95:.4f} × {sigma_male}")
print(f"  x = {x_95_manual:.4f} 英寸")
print(f"\n  ✅ 两种方法结果一致！")
print(f"\n  解读: 95% 的男性身高低于 {x_95:.2f} 英寸。")
````

````python
# ========== 可视化: 第 95 百分位数 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制男性身高分布
x_range = np.linspace(mu_male - 4*sigma_male, mu_male + 4*sigma_male, 1000)
y_range = stats.norm.pdf(x_range, mu_male, sigma_male)

ax.plot(x_range, y_range, 'b-', linewidth=2)

# 填充第 95 百分位数左侧的区域
x_fill = np.linspace(mu_male - 4*sigma_male, x_95, 500)
y_fill = stats.norm.pdf(x_fill, mu_male, sigma_male)
ax.fill_between(x_fill, 0, y_fill, alpha=0.3, color='steelblue',
                label=f'P(X < {x_95:.2f}) = 0.95')

# 标记第 95 百分位数
ax.axvline(x=x_95, color='#e74c3c', linestyle='--', linewidth=2)
ax.text(x_95, -0.01, f'P95 = {x_95:.2f}"', ha='center', fontsize=11, fontweight='bold')

ax.axvline(x=mu_male, color='steelblue', linestyle='--', alpha=0.7, label=f'Mean = {mu_male}"')

ax.set_xlabel('Height (inches)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('95th Percentile of Male Height', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  蓝色区域：身高低于第 95 百分位数的男性（95%）。")
print(f"  红色虚线：第 95 百分位数 = {x_95:.2f} 英寸。")
print(f"  解读：95% 的男性身高低于 {x_95:.2f} 英寸。")
````

## 6. 求分开两端的临界值

### 📐 计算步骤
1. 确定两侧的面积
2. 求对应的 z 分数
3. 还原为原始值

### 📖 书中的例子
> 男性身高服从 $N(68.6, 2.8)$。
> 求分开最低 5% 和最高 5% 的身高值。

````python
# ========== 步骤 5: 求分开两端的临界值 ==========

# 两侧面积
tail_area = 0.05

# 求 z 分数
z_low = stats.norm.ppf(tail_area, 0, 1)
z_high = stats.norm.ppf(1 - tail_area, 0, 1)

# 还原为原始值
x_low = mu_male + z_low * sigma_male
x_high = mu_male + z_high * sigma_male

print(f"\n📊 步骤 5: 求分开最低 5% 和最高 5% 的身高值")
print(f"  男性身高分布: N(μ={mu_male}, σ={sigma_male})")
print(f"  两侧面积: {tail_area}")
print(f"\n  z 分数:")
print(f"    z_{{{tail_area}}} = {z_low:.4f}（左侧 5%）")
print(f"    z_{{{1-tail_area}}} = {z_high:.4f}（右侧 5%）")
print(f"\n  还原为原始值:")
print(f"    x_low = μ + z_low × σ = {mu_male} + ({z_low:.4f}) × {sigma_male} = {x_low:.4f}")
print(f"    x_high = μ + z_high × σ = {mu_male} + {z_high:.4f} × {sigma_male} = {x_high:.4f}")
print(f"\n  解读:")
print(f"    最低 5% 的男性身高低于 {x_low:.2f} 英寸。")
print(f"    最高 5% 的男性身高高于 {x_high:.2f} 英寸。")
print(f"    中间 90% 的男性身高在 {x_low:.2f} 和 {x_high:.2f} 英寸之间。")
````

````python
# ========== 可视化: 分开两端的临界值 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制男性身高分布
x_range = np.linspace(mu_male - 4*sigma_male, mu_male + 4*sigma_male, 1000)
y_range = stats.norm.pdf(x_range, mu_male, sigma_male)

ax.plot(x_range, y_range, 'b-', linewidth=2)

# 填充中间 90% 的区域
x_fill_mid = np.linspace(x_low, x_high, 500)
y_fill_mid = stats.norm.pdf(x_fill_mid, mu_male, sigma_male)
ax.fill_between(x_fill_mid, 0, y_fill_mid, alpha=0.3, color='#2ecc71',
                label=f'Middle 90%')

# 填充左侧 5%
x_fill_left = np.linspace(mu_male - 4*sigma_male, x_low, 500)
y_fill_left = stats.norm.pdf(x_fill_left, mu_male, sigma_male)
ax.fill_between(x_fill_left, 0, y_fill_left, alpha=0.3, color='#e74c3c',
                label=f'Bottom 5%')

# 填充右侧 5%
x_fill_right = np.linspace(x_high, mu_male + 4*sigma_male, 500)
y_fill_right = stats.norm.pdf(x_fill_right, mu_male, sigma_male)
ax.fill_between(x_fill_right, 0, y_fill_right, alpha=0.3, color='#e74c3c',
                label=f'Top 5%')

# 标记临界值
ax.axvline(x=x_low, color='#e74c3c', linestyle='--', linewidth=2)
ax.axvline(x=x_high, color='#e74c3c', linestyle='--', linewidth=2)
ax.axvline(x=mu_male, color='steelblue', linestyle='--', alpha=0.7, label=f'Mean = {mu_male}"')

ax.set_xlabel('Height (inches)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('Male Height: Bottom 5% and Top 5%', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  绿色区域：中间 90% 的男性身高。")
print(f"  红色区域：最低 5% 和最高 5%。")
print(f'  红色虚线：临界值 {x_low:.2f} 和 {x_high:.2f}。')
````

## 7. 质量控制应用

### 📐 范围经验法则
正态分布的数据有以下规律：
- 约 68% 的数据在 $\mu \pm 1\sigma$ 之间
- 约 95% 的数据在 $\mu \pm 2\sigma$ 之间
- 约 99.7% 的数据在 $\mu \pm 3\sigma$ 之间

### 📖 应用场景
> 工厂生产的零件重量服从 $N(50, 2)$ 克。
> 如果零件重量超出 $\mu \pm 2\sigma$ 范围，视为不合格。

````python
# ========== 步骤 6: 质量控制应用 ==========

# 零件重量分布
mu_weight = 50  # 均值（克）
sigma_weight = 2  # 标准差（克）

# 不合格范围
k = 2  # 2 个标准差
lower_limit = mu_weight - k * sigma_weight
upper_limit = mu_weight + k * sigma_weight

# 计算合格率
prob_within = stats.norm.cdf(upper_limit, mu_weight, sigma_weight) - stats.norm.cdf(lower_limit, mu_weight, sigma_weight)
prob_outside = 1 - prob_within

print(f"\n📊 步骤 6: 质量控制应用")
print(f"  零件重量分布: N(μ={mu_weight}, σ={sigma_weight})")
print(f"  不合格范围: 超出 μ ± {k}σ = ({lower_limit}, {upper_limit}) 克")
print(f"\n  合格率: P({lower_limit} < X < {upper_limit}) = {prob_within:.4f}（约 {prob_within*100:.2f}%）")
print(f"  不合格率: P(X < {lower_limit} 或 X > {upper_limit}) = {prob_outside:.4f}（约 {prob_outside*100:.2f}%）")

# 假设生产 10000 个零件
n_parts = 10000
expected_defective = n_parts * prob_outside
print(f"\n  如果生产 {n_parts} 个零件:")
print(f"    预期合格: {n_parts * prob_within:.0f} 个")
print(f"    预期不合格: {expected_defective:.0f} 个")
````

````python
# ========== 可视化: 质量控制 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制零件重量分布
x_range = np.linspace(mu_weight - 4*sigma_weight, mu_weight + 4*sigma_weight, 1000)
y_range = stats.norm.pdf(x_range, mu_weight, sigma_weight)

ax.plot(x_range, y_range, 'b-', linewidth=2)

# 填充合格区域
x_fill_ok = np.linspace(lower_limit, upper_limit, 500)
y_fill_ok = stats.norm.pdf(x_fill_ok, mu_weight, sigma_weight)
ax.fill_between(x_fill_ok, 0, y_fill_ok, alpha=0.3, color='#2ecc71',
                label=f'Within {k}σ: {prob_within:.4f}')

# 填充不合格区域
x_fill_below = np.linspace(mu_weight - 4*sigma_weight, lower_limit, 500)
y_fill_below = stats.norm.pdf(x_fill_below, mu_weight, sigma_weight)
ax.fill_between(x_fill_below, 0, y_fill_below, alpha=0.3, color='#e74c3c')

x_fill_above = np.linspace(upper_limit, mu_weight + 4*sigma_weight, 500)
y_fill_above = stats.norm.pdf(x_fill_above, mu_weight, sigma_weight)
ax.fill_between(x_fill_above, 0, y_fill_above, alpha=0.3, color='#e74c3c',
                label=f'Outside {k}σ: {prob_outside:.4f}')

# 标记控制限
ax.axvline(x=lower_limit, color='#e74c3c', linestyle='--', linewidth=2)
ax.axvline(x=upper_limit, color='#e74c3c', linestyle='--', linewidth=2)
ax.axvline(x=mu_weight, color='steelblue', linestyle='--', alpha=0.7, label=f'Mean = {mu_weight}')

ax.set_xlabel('Weight (grams)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('Quality Control: Part Weight Distribution', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  绿色区域：合格零件（在 {k}σ 范围内）。")
print(f"  红色区域：不合格零件（超出 {k}σ 范围）。")
print(f"  红色虚线：控制限 {lower_limit} 克和 {upper_limit} 克。")
print(f"  约 {prob_within*100:.1f}% 的零件合格。")
````

## 8. 大样本应用

### 📐 计算满足条件的人数
如果知道总体比例，可以计算满足条件的人数。

### 📖 书中的例子
> 假设有 1000 名男性，身高服从 $N(68.6, 2.8)$。
> 求身高在 66 英寸和 72 英寸之间的男性人数。

````python
# ========== 步骤 7: 大样本应用 ==========

# 参数
n_male = 1000
x_low_range = 66
x_high_range = 72

# 计算比例
prob_in_range = stats.norm.cdf(x_high_range, mu_male, sigma_male) - stats.norm.cdf(x_low_range, mu_male, sigma_male)

# 计算人数
n_in_range = n_male * prob_in_range

print(f"\n📊 步骤 7: 大样本应用")
print(f"  男性身高分布: N(μ={mu_male}, σ={sigma_male})")
print(f"  总人数: {n_male}")
print(f"  身高范围: {x_low_range} - {x_high_range} 英寸")
print(f"\n  比例计算:")
print(f"    P({x_low_range} < X < {x_high_range}) = {prob_in_range:.4f}")
print(f"    即约 {prob_in_range*100:.2f}% 的男性在此范围内")
print(f"\n  人数计算:")
print(f"    人数 = 总人数 × 比例 = {n_male} × {prob_in_range:.4f}")
print(f"    人数 = {n_in_range:.0f} 人")
print(f"\n  解读: 在 {n_male} 名男性中，约 {n_in_range:.0f} 人身高在 {x_low_range}-{x_high_range} 英寸之间。")
````

## 📌 核心概念回顾

### 📌 标准化 (Standardization)
- **公式**: $z = \frac{x - \mu}{\sigma}$
- **意义**: 将任意正态分布转换为标准正态分布
- **用途**: 比较不同量纲的数据

### 📌 概率计算
- **给定 x 求概率**: 先标准化为 z，再用 `norm.cdf` 计算
- **给定概率求 x**: 用 `norm.ppf` 求 z，再反标准化
- **两个值之间**: $P(x_1 < X < x_2) = P(Z < z_2) - P(Z < z_1)$

### 📌 范围经验法则
- 约 68% 的数据在 $\mu \pm 1\sigma$ 之间
- 约 95% 的数据在 $\mu \pm 2\sigma$ 之间
- 约 99.7% 的数据在 $\mu \pm 3\sigma$ 之间

### 📌 质量控制
- 控制限通常设为 $\mu \pm 2\sigma$ 或 $\mu \pm 3\sigma$
- 超出控制限的产品视为不合格

### 🔗 完整流程
```
确定分布 → 标准化 → 计算概率/临界值 → 解读结果
    ↓           ↓           ↓               ↓
  N(μ,σ)    z=(x-μ)/σ   cdf 或 ppf      实际意义
```

## ❌ 常见误区

### ❌ 误区 1: 忘记标准化就直接查表
**✓ 正确做法**: 对于非标准正态分布，必须先用 $z = (x - \mu) / \sigma$ 标准化，再查标准正态分布表或用 scipy 计算。

### ❌ 误区 2: 混淆标准化和反标准化
**✓ 正确理解**: 标准化是 $z = (x - \mu) / \sigma$，反标准化是 $x = \mu + z \times \sigma$。两者方向相反。

### ❌ 误区 3: 忘记范围经验法则只是近似
**✓ 正确理解**: 范围经验法则（68-95-99.7）只是近似值，精确概率需要用 scipy 或查表计算。

### ❌ 误区 4: 把总体参数和样本统计量混淆
**✓ 正确理解**: 标准化公式中的 $\mu$ 和 $\sigma$ 是**总体参数**，不是样本均值和标准差。

### ❌ 误区 5: 在样本量小时用正态分布代替 t 分布
**✓ 正确理解**: 当总体标准差未知且样本量小（$n < 30$）时，应该用 t 分布而不是正态分布。
