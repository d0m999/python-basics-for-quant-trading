## 1. 场景设定：为什么需要概率？

### 🎯 一个直觉问题

Microsoft XSORT 性别选择技术的设计目的是增加新生儿是女孩的可能性。在临床试验中，945 个新生儿中有 879 个女孩。

**核心问题**：如果 20 对夫妇使用该方法，20 个新生儿都是女孩的概率是多少？这个结果能否证明该方法有效？

### 📖 书中的定义

> **事件**是一个过程的结果或结果的任意集合。
> **简单事件**是不能进一步拆分的结果或事件。
> **样本空间**由所有可能的简单事件组成。

概率论是推断统计的基础——假设检验中的 p 值就是一个概率值。

## 2. 样本空间与简单事件

### 📐 概念定义

以新生儿性别为例：

| 实验 | 样本空间 | 简单事件数 |
|------|----------|------------|
| 1 个新生儿 | {b, g} | 2 |
| 2 个新生儿 | {bb, bg, gb, gg} | 4 |
| 3 个新生儿 | {bbb, bbg, bgb, bgg, gbb, gbg, ggb, ggg} | 8 |

其中 b = 男孩，g = 女孩。

**关键观察**：3 个新生儿的样本空间有 $2^3 = 8$ 个简单事件。

````python
# ========== 步骤 1: 构建样本空间 ==========

# 1 个新生儿
space_1 = ['b', 'g']
print("📊 步骤 1: 构建样本空间")
print(f"\n  1 个新生儿的样本空间: {space_1}")
print(f"  简单事件数: {len(space_1)}")

# 2 个新生儿
space_2 = [''.join(p) for p in product('bg', repeat=2)]
print(f"\n  2 个新生儿的样本空间: {space_2}")
print(f"  简单事件数: {len(space_2)}")

# 3 个新生儿
space_3 = [''.join(p) for p in product('bg', repeat=3)]
print(f"\n  3 个新生儿的样本空间: {space_3}")
print(f"  简单事件数: {len(space_3)}")

# 验证公式: 2^n
print(f"\n💡 验证: 2^1={2**1}, 2^2={2**2}, 2^3={2**3}")
print(f"  样本空间大小 = 2^(新生儿数)")
````

````python
# ========== 事件的识别 ==========

# 在 3 个新生儿中，"2 女 1 男" 是一个复合事件
event_2g1b = [s for s in space_3 if s.count('g') == 2]
print("📊 复合事件示例: 3 个新生儿中 '2 女 1 男'")
print(f"  包含的简单事件: {event_2g1b}")
print(f"  简单事件数: {len(event_2g1b)}")

# "性别相同" 事件
event_same = [s for s in space_3 if s.count('g') == 3 or s.count('b') == 3]
print(f"\n📊 复合事件示例: 3 个新生儿 '性别相同'")
print(f"  包含的简单事件: {event_same}")
print(f"  简单事件数: {len(event_same)}")
````

## 3. 计算概率的三种方法

### 📐 三种方法概览

| 方法 | 适用条件 | 公式 |
|------|----------|------|
| 相对频数法 | 有历史数据或可重复实验 | $P(A) \approx \frac{\text{事件A发生的次数}}{\text{重复实验的次数}}$ |
| 经典计算法 | 结果等可能 | $P(A) = \frac{s}{n}$（s = 有利结果数，n = 总结果数） |
| 主观估计法 | 无数据、不可重复 | 根据经验估计 |

> ⚠️ 使用经典计算法时，务必确认结果是等可能性的！

### 3.1 经典计算法示例

````python
# ========== 步骤 2: 经典计算法 - 3 个性别相同的概率 ==========

# 样本空间: 8 个等可能的简单事件
total_outcomes = len(space_3)
favorable = len(event_same)

prob_same_gender = favorable / total_outcomes

print("📊 步骤 2: 经典计算法")
print(f"  样本空间大小 n = {total_outcomes}")
print(f"  有利结果数 s = {favorable} (bbb, ggg)")
print(f"  计算公式: P(A) = s/n")
print(f"  实际计算: {favorable}/{total_outcomes} = {prob_same_gender:.4f}")
print(f"\n💡 3 个新生儿性别相同的概率为 {prob_same_gender:.2f}，即 25%")
````

### 3.2 相对频数法：模拟掷骰子

````python
# ========== 步骤 3: 相对频数法 - 掷骰子模拟 ==========

# 模拟掷一个公平骰子
print("📊 步骤 3: 相对频数法 - 掷骰子模拟")
print("  目标: 估计掷出 6 的概率")
print(f"  经典计算法答案: P(6) = 1/6 ≈ {1/6:.4f}")

# 不同样本量下的相对频数
sample_sizes = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]
results = []

for n in sample_sizes:
    rolls = np.random.randint(1, 7, size=n)
    freq_6 = np.sum(rolls == 6) / n
    results.append({'样本量': n, '相对频数': freq_6, '偏差': abs(freq_6 - 1/6)})

df_results = pd.DataFrame(results)
print("\n" + df_results.to_string(index=False))
print(f"\n💡 大数定律: 随着样本量增加，相对频数趋近于真实概率 {1/6:.4f}")
````

````python
# ========== 可视化: 大数定律 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 逐步累积的相对频数
n_trials = 10000
rolls = np.random.randint(1, 7, size=n_trials)
cumulative_freq_6 = np.cumsum(rolls == 6) / np.arange(1, n_trials + 1)

ax.plot(range(1, n_trials + 1), cumulative_freq_6, color='steelblue', linewidth=0.8, alpha=0.8)
ax.axhline(y=1/6, color='#e74c3c', linestyle='--', linewidth=2, label='True Probability = 1/6')
ax.set_xlabel('Number of Trials', fontsize=12)
ax.set_ylabel('Relative Frequency of 6', fontsize=12)
ax.set_title('Law of Large Numbers: Rolling a Fair Die', fontsize=14, fontweight='bold')
ax.legend(fontsize=12)
ax.set_ylim(0, 0.3)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\n💡 图解说明：")
print("  - 蓝线: 累积相对频数")
print("  - 红色虚线: 真实概率 1/6 ≈ 0.1667")
print("  - 随着试验次数增加，相对频数越来越接近真实概率")
print("  - 这就是大数定律的直观体现")
````

### 3.3 相对频数法：实际案例

````python
# ========== 步骤 4: 相对频数法 - 空难概率 ==========

# 书中例题: 某年约 3900 万个商业航班，空难 16 次
total_flights = 39_000_000
crashes = 16

prob_crash = crashes / total_flights

print("📊 步骤 4: 相对频数法 - 空难概率")
print(f"  总航班数 = {total_flights:,}")
print(f"  空难次数 = {crashes}")
print(f"  计算公式: P(A) = 事件A发生次数 / 总次数")
print(f"  实际计算: {crashes}/{total_flights:,} = {prob_crash:.10f}")
print(f"  即约 {prob_crash*1000000:.4f} 百万分之一")
print(f"\n💡 该概率属于 FAA 定义的 '极不可能' 级别 (< 0.000000001)")
````

````python
# ========== 步骤 5: 相对频数法 - 调查数据 ==========

# 书中例题: 皮尤研究中心调查 - 是否见过鬼
yes_count = 366
no_count = 1637
total_count = yes_count + no_count

prob_ghost = yes_count / total_count

print("📊 步骤 5: 相对频数法 - 调查数据")
print(f"  回答 '是' 的人数 = {yes_count}")
print(f"  回答 '否' 的人数 = {no_count}")
print(f"  总回答数 = {total_count}")
print(f"\n  ⚠️ 常见错误: 直接用 366/1637 = {366/1637:.3f}")
print(f"  ✓ 正确计算: 366/{total_count} = {prob_ghost:.3f}")
print(f"\n💡 P(见过鬼) = {prob_ghost:.3f}，注意分母是总回答数，不是回答'否'的人数")
````

## 4. 概率的范围与小概率原理

### 📐 概率的基本性质

$$0 \leq P(A) \leq 1$$

| 概率值 | 含义 |
|--------|------|
| 0 | 不可能发生 |
| 接近 0 | 极不可能 |
| 0.5 | 50-50 可能 |
| 接近 1 | 非常可能 |
| 1 | 一定发生 |

### 📐 小概率原理（稀有事件规则）

> 如果在给定的假设下，某一事件发生的概率非常低（$\leq 0.05$），则可以得出结论：该假设可能不正确。

- **显著高的成功次数**：$P(x \text{ 或更多}) \leq 0.05$
- **显著低的成功次数**：$P(x \text{ 或更少}) \leq 0.05$

````python
# ========== 步骤 6: 小概率原理 - XSORT 性别选择 ==========

# 书中例题: 20 个新生儿全是女孩的概率
# 假设方法无效，P(女孩) = 0.5

n_births = 20
p_girl = 0.5

# 20 个全是女孩的概率 (独立事件，用乘法原理)
prob_all_girls = p_girl ** n_births

print("📊 步骤 6: 小概率原理 - XSORT 性别选择")
print(f"  假设: XSORT 方法无效，P(女孩) = 0.5")
print(f"  观测: 20 个新生儿全是女孩")
print(f"\n  计算公式: P(20个女孩) = 0.5^20")
print(f"  实际计算: {prob_all_girls:.10f}")
print(f"  即约 {1/prob_all_girls:,.0f} 分之一")

# 用 scipy.stats.binom 验证
prob_binom = stats.binom.pmf(20, 20, 0.5)
print(f"\n🔬 scipy.stats.binom 验证:")
print(f"  手算概率 = {prob_all_girls:.10f}")
print(f"  scipy 概率 = {prob_binom:.10f}")
print(f"\n  ✅ 验证通过！")

# 判断显著性
alpha = 0.05
print(f"\n🎯 结论:")
print(f"  P(20个女孩) = {prob_all_girls:.10f} < α = {alpha}")
print(f"  → 这是一个显著高的结果")
print(f"  → 应排除偶然发生的可能性")
print(f"  → XSORT 方法是有效的")
````

````python
# ========== 对比: 20 个新生儿中有 12 个女孩 ==========

# P(12个或更多女孩)
prob_12_or_more = 1 - stats.binom.cdf(11, 20, 0.5)

print("📊 对比: 12 个女孩是否显著？")
print(f"  P(12个或更多女孩) = {prob_12_or_more:.6f}")
print(f"  α = {alpha}")
print(f"\n  P = {prob_12_or_more:.6f} > α = {alpha}")
print(f"  → 这不是一个显著高的结果")
print(f"  → 不能排除偶然发生的可能性")
print(f"  → 无法证明 XSORT 方法有效")
````

````python
# ========== 可视化: 二项分布与显著性 ==========

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

x = np.arange(0, 21)
probs = stats.binom.pmf(x, 20, 0.5)

# 左图: 20 个女孩
colors = ['#2ecc71' if xi < 20 else '#e74c3c' for xi in x]
axes[0].bar(x, probs, color=colors, edgecolor='white', alpha=0.8)
axes[0].axhline(y=0.05, color='red', linestyle='--', alpha=0.5, label='α = 0.05')
axes[0].set_xlabel('Number of Girls', fontsize=12)
axes[0].set_ylabel('Probability', fontsize=12)
axes[0].set_title('P(20 girls) = 0.000000954 - Significant!', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(axis='y', alpha=0.3)

# 右图: 12 个或更多女孩
colors2 = ['#2ecc71' if xi < 12 else '#e74c3c' for xi in x]
axes[1].bar(x, probs, color=colors2, edgecolor='white', alpha=0.8)
axes[1].axhline(y=0.05, color='red', linestyle='--', alpha=0.5, label='α = 0.05')
axes[1].set_xlabel('Number of Girls', fontsize=12)
axes[1].set_ylabel('Probability', fontsize=12)
axes[1].set_title(f'P(≥12 girls) = {prob_12_or_more:.4f} - Not Significant', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n💡 图解说明：")
print("  左图: 20 个女孩的概率极低（红色），远低于 α=0.05，结果显著")
print("  右图: 12+ 个女孩的概率（红色区域）高于 α=0.05，结果不显著")
print("  → 统计学家根据低概率排除偶然发生的可能性")
````

## 5. 对立事件

### 📐 对立事件准则

事件 A 的对立事件记为 $\bar{A}$，由事件 A 未发生的所有结果组成。

$$P(A) + P(\bar{A}) = 1$$
$$P(A) = 1 - P(\bar{A})$$
$$P(\bar{A}) = 1 - P(A)$$

> 💡 当直接计算 $P(A)$ 困难时，可以先计算 $P(\bar{A})$，再用 $P(A) = 1 - P(\bar{A})$

````python
# ========== 步骤 7: 对立事件 ==========

# 书中例题: 皮尤研究中心调查 - 是否上网
internet_yes = 1782
total_survey = 2002
internet_no = total_survey - internet_yes

prob_internet = internet_yes / total_survey
prob_no_internet = internet_no / total_survey
prob_no_internet_verify = 1 - prob_internet

print("📊 步骤 7: 对立事件 - 上网概率")
print(f"  调查人数 = {total_survey}")
print(f"  上网人数 = {internet_yes}")
print(f"  不上网人数 = {internet_no}")
print(f"\n  P(上网) = {internet_yes}/{total_survey} = {prob_internet:.3f}")
print(f"  P(不上网) = {internet_no}/{total_survey} = {prob_no_internet:.3f}")
print(f"\n  验证: P(不上网) = 1 - P(上网) = 1 - {prob_internet:.3f} = {prob_no_internet_verify:.3f}")
print(f"\n💡 对立事件准则: P(A) + P(Ā) = 1")
````

## 6. 模拟实验：生日问题

### 🎯 经典问题

在全班 25 名学生中，至少有两名学生同一天生日（不同年份）的概率超过 50%！

### 📐 计算方法

使用对立事件：
$$P(\text{至少两人同天生日}) = 1 - P(\text{所有人生日都不同})$$

$$P(\text{所有人生日都不同}) = \frac{365}{365} \times \frac{364}{365} \times \frac{363}{365} \times \cdots \times \frac{365-n+1}{365}$$

````python
# ========== 步骤 8: 生日问题 - 理论计算 ==========

def birthday_prob(n):
    """计算 n 人中至少两人同天生日的概率"""
    prob_all_different = 1.0
    for i in range(n):
        prob_all_different *= (365 - i) / 365
    return 1 - prob_all_different

print("📊 步骤 8: 生日问题 - 理论计算")
print(f"  {'人数':>4}  {'至少两人同天生日的概率':>25}")
print(f"  {'-'*4}  {'-'*25}")

for n in [2, 5, 10, 15, 20, 23, 25, 30, 40, 50, 60, 70, 80, 100]:
    p = birthday_prob(n)
    marker = " ← 超过 50%!" if n == 23 else ""
    print(f"  {n:>4}  {p:>25.6f}{marker}")

print(f"\n💡 23 人时概率就超过 50%，这与直觉相差很大！")
````

````python
# ========== 步骤 9: 生日问题 - 模拟验证 ==========

n_simulations = 10000
n_people = 23

count_match = 0
for _ in range(n_simulations):
    birthdays = np.random.randint(0, 365, size=n_people)
    if len(birthdays) != len(set(birthdays)):
        count_match += 1

sim_prob = count_match / n_simulations
theory_prob = birthday_prob(n_people)

print("📊 步骤 9: 生日问题 - 模拟验证")
print(f"  人数: {n_people}")
print(f"  模拟次数: {n_simulations:,}")
print(f"  有匹配的次数: {count_match:,}")
print(f"\n  模拟概率 = {sim_prob:.4f}")
print(f"  理论概率 = {theory_prob:.4f}")
print(f"  偏差 = {abs(sim_prob - theory_prob):.4f}")
print(f"\n  ✅ 模拟结果与理论值接近！")
````

````python
# ========== 可视化: 生日问题 ==========

fig, ax = plt.subplots(figsize=(12, 6))

n_range = np.arange(2, 81)
probs = [birthday_prob(n) for n in n_range]

ax.plot(n_range, probs, 'b-', linewidth=2, label='Theoretical Probability')
ax.axhline(y=0.5, color='#e74c3c', linestyle='--', alpha=0.7, label='50% threshold')
ax.axvline(x=23, color='#e67e22', linestyle='--', alpha=0.7, label='n = 23')
ax.fill_between(n_range, 0, probs, alpha=0.1, color='steelblue')

ax.set_xlabel('Number of People (n)', fontsize=12)
ax.set_ylabel('P(at least 2 share birthday)', fontsize=12)
ax.set_title('Birthday Problem: Probability vs Group Size', fontsize=14, fontweight='bold')
ax.legend(fontsize=12)
ax.grid(alpha=0.3)
ax.set_ylim(0, 1.05)

plt.tight_layout()
plt.show()

print("\n💡 图解说明：")
print("  - 蓝线: 至少两人同天生日的概率")
print("  - 红线: 50% 阈值")
print("  - 橙线: n=23，此时概率首次超过 50%")
print("  - n=50 时概率已达 97%，n=70 时超过 99.9%")
````

## 7. 大规模模拟：验证三种方法的一致性

````python
# ========== 步骤 10: 抽牌模拟 - 验证经典计算法 ==========

# 从一副标准 52 张牌中抽到红心的概率
# 经典计算法: 13/52 = 0.25

n_draws = 100000
deck = ['红心']*13 + ['方块']*13 + ['梅花']*13 + ['黑桃']*13

# 模拟抽牌
draws = np.random.choice(deck, size=n_draws, replace=True)
freq_hearts = np.sum(draws == '红心') / n_draws

print("📊 步骤 10: 抽牌模拟 - 验证经典计算法")
print(f"  目标: 抽到红心的概率")
print(f"  经典计算法: P(红心) = 13/52 = {13/52:.4f}")
print(f"  相对频数法: P(红心) ≈ {freq_hearts:.4f} (基于 {n_draws:,} 次模拟)")
print(f"  偏差: {abs(freq_hearts - 13/52):.4f}")
print(f"\n  ✅ 两种方法结果一致！")
````

````python
# ========== 步骤 11: 掷两个骰子 - 点数之和的分布 ==========

# 理论分布
theoretical = {}
for d1 in range(1, 7):
    for d2 in range(1, 7):
        s = d1 + d2
        theoretical[s] = theoretical.get(s, 0) + 1

for k in theoretical:
    theoretical[k] /= 36

# 模拟
n_rolls = 100000
die1 = np.random.randint(1, 7, size=n_rolls)
die2 = np.random.randint(1, 7, size=n_rolls)
sums = die1 + die2

print("📊 步骤 11: 掷两个骰子 - 点数之和")
print(f"  {'点数和':>6}  {'理论概率':>10}  {'模拟概率':>10}  {'偏差':>10}")
print(f"  {'-'*6}  {'-'*10}  {'-'*10}  {'-'*10}")

for s in range(2, 13):
    theory_p = theoretical[s]
    sim_p = np.sum(sums == s) / n_rolls
    print(f"  {s:>6}  {theory_p:>10.4f}  {sim_p:>10.4f}  {abs(theory_p - sim_p):>10.4f}")

print(f"\n💡 经典计算法和相对频数法的结果高度一致")
````

````python
# ========== 可视化: 掷两个骰子的分布 ==========

fig, ax = plt.subplots(figsize=(12, 6))

sums_range = range(2, 13)
theory_probs = [theoretical[s] for s in sums_range]
sim_probs = [np.sum(sums == s) / n_rolls for s in sums_range]

x = np.arange(len(sums_range))
width = 0.35

bars1 = ax.bar(x - width/2, theory_probs, width, label='Theoretical (Classical)', 
               color='steelblue', alpha=0.8, edgecolor='white')
bars2 = ax.bar(x + width/2, sim_probs, width, label='Simulated (Relative Frequency)', 
               color='#2ecc71', alpha=0.8, edgecolor='white')

ax.set_xlabel('Sum of Two Dice', fontsize=12)
ax.set_ylabel('Probability', fontsize=12)
ax.set_title('Two Dice: Theoretical vs Simulated Probabilities', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(sums_range)
ax.legend(fontsize=12)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n💡 图解说明：")
print("  蓝色: 经典计算法的理论概率")
print("  绿色: 100,000 次模拟的相对频数")
print("  两者几乎完全重合，验证了两种方法的一致性")
print("  7 是最可能的点数和（6/36 ≈ 0.167），2 和 12 最不可能（1/36 ≈ 0.028）")
````

## 8. 发生比（Odds）

### 📐 发生比的定义

| 概念 | 公式 |
|------|------|
| 实际劣势发生比 | $P(\bar{A}) : P(A)$ |
| 实际优势发生比 | $P(A) : P(\bar{A})$ |
| 赔率 | 净利润 : 赌注 |

> 💡 概率使计算更简便（统计学家偏好），发生比在博彩中更常用。

````python
# ========== 步骤 12: 发生比 - 轮盘赌示例 ==========

# 书中例题: 轮盘赌数字 13
# P(A) = 1/38, P(Ā) = 37/38
# 赌场赔率 = 35:1

p_A = 1/38
p_not_A = 37/38

# 实际劣势发生比
odds_against = p_not_A / p_A

# 赌场赔率
casino_odds = 35
bet = 5  # 美元

print("📊 步骤 12: 发生比 - 轮盘赌")
print(f"  事件 A: 轮盘停在数字 13")
print(f"  P(A) = 1/38 = {p_A:.4f}")
print(f"  P(Ā) = 37/38 = {p_not_A:.4f}")
print(f"\n  实际劣势发生比 = P(Ā)/P(A) = {odds_against:.0f}:1")
print(f"  赌场赔率 = {casino_odds}:1")
print(f"\n  如果赌 {bet} 美元赢了:")
print(f"    赌场赔率下净利润 = {bet} × {casino_odds} = ${bet * casino_odds}")
print(f"    实际发生比下净利润 = {bet} × {odds_against:.0f} = ${bet * odds_against:.0f}")
print(f"\n💡 赌场赔率(35:1) < 实际发生比(37:1)，这就是赌场盈利的方式")
````

## 📌 核心概念回顾

### 📌 概率 (Probability)
- **定义**: 事件发生可能性的数值度量
- **范围**: $0 \leq P(A) \leq 1$
- **计算方法**: 相对频数法、经典计算法、主观估计法

### 📌 样本空间 (Sample Space)
- **定义**: 所有可能的简单事件的集合
- **公式**: n 个二元试验的样本空间大小 = $2^n$

### 📌 对立事件 (Complement)
- **公式**: $P(A) + P(\bar{A}) = 1$
- **应用**: 当 $P(A)$ 难以直接计算时，用 $P(A) = 1 - P(\bar{A})$

### 📌 小概率原理 (Rare Event Rule)
- **定义**: 如果某事件概率 $\leq 0.05$，则认为该假设可能不正确
- **应用**: 假设检验的基础

### 📌 大数定律 (Law of Large Numbers)
- **定义**: 多次重复实验，相对频数趋近于真实概率
- **含义**: 少数实验可能有偏差，大量实验则更准确

### 🔗 完整流程
```
定义事件 → 确定样本空间 → 选择计算方法 → 计算概率 → 判断显著性
    ↓            ↓              ↓            ↓           ↓
  A是啥     所有可能结果    有数据?等可能?   P(A)的值    P≤0.05?
```

## ❌ 常见误区

### ❌ 误区 1: 不等可能事件误用经典计算法
**✓ 正确理解**: 经典计算法要求所有简单事件等可能。例如，不能认为考试及格的概率是 0.5（及格或不及格），实际概率取决于复习程度和考试难度。

### ❌ 误区 2: 分母选择错误
**✓ 正确理解**: 计算概率时，分母是总回答数/总试验数，不是某个子集的大小。例如，366 人回答「是」、1637 人回答「否」，P(是) = 366/2003，不是 366/1637。

### ❌ 误区 3: 混淆统计显著性与实际重要性
**✓ 正确理解**: 一个结果统计上显著（概率低），不代表它在实际中很重要。统计显著性只说明不太可能是偶然发生的。

### ❌ 误区 4: 赌徒谬误
**✓ 正确理解**: 大数定律适用于大量实验，不适用于单次实验。一连串的输钱不会增加下一次赢钱的概率，独立事件没有「记忆」。

### ❌ 误区 5: 将 P(B|A) 与 P(A|B) 混淆
**✓ 正确理解**: 条件概率通常不对称。例如，P(阳性|患病) ≠ P(患病|阳性)。这将在 4-3 节详细讨论。
