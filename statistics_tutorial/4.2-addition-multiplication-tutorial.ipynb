## 1. 场景设定：复合事件的概率

### 🎯 核心问题

4-1 节只涉及简单事件。实际中我们常需要计算：
- 事件 A **或**事件 B 发生的概率 → 加法原理（"或" = 加法）
- 事件 A **和**事件 B 都发生的概率 → 乘法原理（"和" = 乘法）

### 📖 书中的定义

> **复合事件**是由两个或多个简单事件组合而成的事件。
> **互斥事件**（互不相容事件）：事件 A 和事件 B 不能同时发生。
> **独立事件**：事件 A 的发生不影响事件 B 发生的概率。

## 2. 加法原理

### 📐 加法原理公式

$$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$

其中：
- $P(A \cup B)$ = 事件 A 发生或事件 B 发生（或两者都发生）的概率
- $P(A \cap B)$ = 事件 A 和事件 B 同时发生的概率
- $\cup$ 表示"并集"（或），$\cap$ 表示"交集"（和）

> 💡 直觉：把 A 和 B 的有利结果相加，但重叠部分只算一次，所以要减去。

````python
# ========== 步骤 1: 加法原理 - 毒品检验数据 ==========

# 书中表 4-2: 毒品检验结果
#                  检验阳性    检验阴性    合计
# 吸食毒品           45          5         50
# 未吸食毒品         25         480       505
# 合计               70         485       555

total = 555
positive = 70      # 检验阳性总数
drugs = 50         # 吸食毒品总数
positive_and_drugs = 45  # 阳性且吸食毒品

# 加法原理
prob_positive = positive / total
prob_drugs = drugs / total
prob_both = positive_and_drugs / total
prob_positive_or_drugs = prob_positive + prob_drugs - prob_both

# 直接计数验证
positive_or_drugs_count = 45 + 25 + 5  # 阳性或吸食毒品的人数
prob_direct = positive_or_drugs_count / total

print("📊 步骤 1: 加法原理 - 毒品检验数据")
print(f"  总人数 = {total}")
print(f"  检验阳性人数 = {positive}")
print(f"  吸食毒品人数 = {drugs}")
print(f"  阳性且吸食毒品人数 = {positive_and_drugs}")
print(f"\n  计算公式: P(A∪B) = P(A) + P(B) - P(A∩B)")
print(f"  实际计算: {prob_positive:.4f} + {prob_drugs:.4f} - {prob_both:.4f} = {prob_positive_or_drugs:.4f}")
print(f"\n  直接计数验证: {positive_or_drugs_count}/{total} = {prob_direct:.4f}")
print(f"\n  ✅ 验证通过！")
print(f"\n💡 为什么不能直接相加? 因为 45 人同时满足两个条件，会被重复计算")
````

### 📐 互斥事件的加法规则

当事件 A 和事件 B 为互斥事件时，$P(A \cap B) = 0$，因此：

$$P(A \cup B) = P(A) + P(B)$$

````python
# ========== 步骤 2: 互斥事件示例 ==========

# 互斥事件: 随机选取男性 vs 女性参加临床试验
# 非互斥事件: 上统计课的学生 vs 女性学生

print("📊 步骤 2: 互斥事件 vs 非互斥事件")
print(f"\n  互斥事件示例:")
print(f"    事件 A: 随机选取男性受试者")
print(f"    事件 B: 随机选取女性受试者")
print(f"    → 一个人不可能同时是男性和女性")
print(f"    → P(A∩B) = 0")
print(f"    → P(A∪B) = P(A) + P(B)")

print(f"\n  非互斥事件示例:")
print(f"    事件 A: 随机选取上统计课的学生")
print(f"    事件 B: 随机选取女性学生")
print(f"    → 一个人可以同时上统计课且是女性")
print(f"    → P(A∩B) ≠ 0")
print(f"    → P(A∪B) = P(A) + P(B) - P(A∩B)")
````

## 3. 对立事件准则

### 📐 对立事件与加法原理

事件 A 和其对立事件 $\bar{A}$ 一定是互斥的，且 $P(A \cup \bar{A}) = 1$：

$$P(A) + P(\bar{A}) = 1$$
$$P(A) = 1 - P(\bar{A})$$

````python
# ========== 步骤 3: 对立事件准则 ==========

# 书中例题: 智能手机拥有率
prob_smartphone = 0.87
prob_no_smartphone = 1 - prob_smartphone

print("📊 步骤 3: 对立事件准则")
print(f"  P(有智能手机) = {prob_smartphone}")
print(f"  P(没有智能手机) = 1 - {prob_smartphone} = {prob_no_smartphone}")
print(f"\n  验证: P(A) + P(Ā) = {prob_smartphone} + {prob_no_smartphone} = {prob_smartphone + prob_no_smartphone}")
print(f"\n💡 对立事件准则是简化计算的利器：当 P(A) 难算时，先算 P(Ā)")
````

## 4. 乘法原理

### 📐 乘法原理公式

$$P(A \cap B) = P(A) \times P(B|A)$$

其中：
- $P(A \cap B)$ = 事件 A 发生并且事件 B 发生的概率
- $P(B|A)$ = 在事件 A 已经发生的条件下，事件 B 发生的概率（条件概率）

> 💡 直觉：先算 A 发生的概率，再算在 A 已发生的情况下 B 发生的概率，两者相乘。

### 📐 独立事件与非独立事件

| 类型 | 定义 | 乘法规则 |
|------|------|----------|
| 独立事件 | P(B|A) = P(B) | P(A∩B) = P(A) × P(B) |
| 非独立事件 | P(B|A) ≠ P(B) | P(A∩B) = P(A) × P(B|A) |

> ⚠️ 非独立事件不意味着一个是另一个的直接原因！

````python
# ========== 步骤 4: 乘法原理 - 有放回 vs 无放回 ==========

# 书中例题: 毒品检验受试者中抽取 2 人
# 50 名吸食毒品的受试者: 45 人阳性, 5 人阴性

n_total = 50
n_positive = 45
n_negative = 5

# 有放回: 独立事件
prob_pos_with_replace = (n_positive / n_total) * (n_negative / n_total)

# 无放回: 非独立事件
# 第一次抽到阳性后，剩下 49 人，其中 5 人阴性
prob_pos_without_replace = (n_positive / n_total) * (n_negative / (n_total - 1))

print("📊 步骤 4: 乘法原理 - 有放回 vs 无放回")
print(f"  50 名受试者: 45 人阳性, 5 人阴性")
print(f"  目标: 第一次抽到阳性，第二次抽到阴性")
print(f"\n  有放回（独立事件）:")
print(f"    P(第1次阳性) = 45/50 = {n_positive/n_total:.4f}")
print(f"    P(第2次阴性) = 5/50 = {n_negative/n_total:.4f}")
print(f"    P(阳性∩阴性) = 45/50 × 5/50 = {prob_pos_with_replace:.4f}")
print(f"\n  无放回（非独立事件）:")
print(f"    P(第1次阳性) = 45/50 = {n_positive/n_total:.4f}")
print(f"    P(第2次阴性|第1次阳性) = 5/49 = {n_negative/(n_total-1):.4f}")
print(f"    P(阳性∩阴性) = 45/50 × 5/49 = {prob_pos_without_replace:.4f}")
print(f"\n  差异: {abs(prob_pos_without_replace - prob_pos_with_replace):.4f}")
print(f"\n💡 有放回→独立事件，无放回→非独立事件")
````

````python
# ========== 步骤 5: 模拟验证乘法原理 ==========

n_simulations = 100000

# 构建总体
population = ['阳性'] * n_positive + ['阴性'] * n_negative

# 有放回模拟
count_with = 0
for _ in range(n_simulations):
    draw1 = np.random.choice(population)
    draw2 = np.random.choice(population)
    if draw1 == '阳性' and draw2 == '阴性':
        count_with += 1
sim_with = count_with / n_simulations

# 无放回模拟
count_without = 0
for _ in range(n_simulations):
    idx1 = np.random.randint(0, len(population))
    remaining = population[:idx1] + population[idx1+1:]
    idx2 = np.random.randint(0, len(remaining))
    if population[idx1] == '阳性' and remaining[idx2] == '阴性':
        count_without += 1
sim_without = count_without / n_simulations

print("📊 步骤 5: 模拟验证乘法原理")
print(f"  模拟次数: {n_simulations:,}")
print(f"\n  有放回:")
print(f"    理论概率 = {prob_pos_with_replace:.4f}")
print(f"    模拟概率 = {sim_with:.4f}")
print(f"    偏差 = {abs(sim_with - prob_pos_with_replace):.4f}")
print(f"\n  无放回:")
print(f"    理论概率 = {prob_pos_without_replace:.4f}")
print(f"    模拟概率 = {sim_without:.4f}")
print(f"    偏差 = {abs(sim_without - prob_pos_without_replace):.4f}")
print(f"\n  ✅ 模拟结果与理论值一致！")
````

## 5. 5% 烦琐计算准则

### 📐 准则内容

> 如果考虑无放回抽样并且样本量不超过总体大小的 5%，那么可以将抽样视为相互独立的（即使实际上并不独立）。

这样可以避免复杂的条件概率计算。

````python
# ========== 步骤 6: 5% 准则 - 扩展到 3 次抽样 ==========

# 书中例题: 从 130,639,273 名全职工作者中抽取 3 人
# P(阳性) = 0.042

n_population = 130_639_273
p_positive = 0.042
n_sample = 3

# 检查是否满足 5% 准则
ratio = n_sample / n_population

# 使用 5% 准则（视为独立事件）
prob_3_positive_independent = p_positive ** 3

# 不使用 5% 准则（非独立事件）
n_positive_pop = int(n_population * p_positive)
prob_3_positive_dependent = 1.0
for i in range(3):
    prob_3_positive_dependent *= (n_positive_pop - i) / (n_population - i)

print("📊 步骤 6: 5% 准则 - 扩展到 3 次抽样")
print(f"  总体大小: {n_population:,}")
print(f"  样本量: {n_sample}")
print(f"  样本/总体 = {ratio:.2e} << 0.05")
print(f"  → 满足 5% 准则，可视为独立事件")
print(f"\n  独立事件计算: P(3人阳性) = 0.042^3 = {prob_3_positive_independent:.10f}")
print(f"  非独立事件计算: P(3人阳性) = {prob_3_positive_dependent:.10f}")
print(f"  偏差: {abs(prob_3_positive_dependent - prob_3_positive_independent):.10f}")
print(f"\n💡 5% 准则让计算大大简化，且误差可以忽略")
````

## 6. 冗余设计：乘法原理的重要应用

### 📐 系统可靠性

如果一个部件的失效率为 0.001，两个独立部件同时失效的概率仅为 $0.001^2 = 0.000001$。

$$P(\text{系统故障}) = P(\text{部件1故障}) \times P(\text{部件2故障})$$

````python
# ========== 步骤 7: 冗余设计 - 硬盘故障率 ==========

# 书中例题: 希捷硬盘一年内故障率 2.89%
failure_rate = 0.0289
success_rate = 1 - failure_rate

# 一块硬盘正常工作的概率
prob_1_drive = success_rate

# 两块独立硬盘至少一块正常工作的概率
prob_both_fail = failure_rate ** 2
prob_at_least_1 = 1 - prob_both_fail

print("📊 步骤 7: 冗余设计 - 硬盘故障率")
print(f"  单块硬盘故障率 = {failure_rate:.4f} ({failure_rate*100:.2f}%)")
print(f"  单块硬盘正常率 = {success_rate:.4f} ({success_rate*100:.2f}%)")
print(f"\n  一块硬盘: P(正常) = {prob_1_drive:.4f}")
print(f"  两块硬盘: P(至少一块正常) = 1 - P(两块都故障)")
print(f"           = 1 - {failure_rate}^2 = 1 - {prob_both_fail:.6f} = {prob_at_least_1:.6f}")
print(f"\n  故障风险降低倍数: {failure_rate/prob_both_fail:.1f} 倍")
print(f"\n💡 冗余设计是乘法原理的经典应用：独立部件同时故障的概率极低")
````

````python
# ========== 可视化: 冗余设计的效果 ==========

fig, ax = plt.subplots(figsize=(12, 6))

n_drives = range(1, 6)
system_failure = [failure_rate ** n for n in n_drives]
system_success = [1 - f for f in system_failure]

x = np.arange(len(n_drives))
width = 0.35

bars1 = ax.bar(x - width/2, system_failure, width, label='System Failure Rate',
               color='#e74c3c', alpha=0.8, edgecolor='white')
bars2 = ax.bar(x + width/2, system_success, width, label='System Success Rate',
               color='#2ecc71', alpha=0.8, edgecolor='white')

ax.set_xlabel('Number of Redundant Drives', fontsize=12)
ax.set_ylabel('Probability', fontsize=12)
ax.set_title('Redundancy Design: Failure Rate Reduction', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(n_drives)
ax.legend(fontsize=12)
ax.grid(axis='y', alpha=0.3)
ax.set_yscale('log')

for i, (f, s) in enumerate(zip(system_failure, system_success)):
    ax.text(i - width/2, f, f'{f:.2e}', ha='center', va='bottom', fontsize=9)
    ax.text(i + width/2, s, f'{s:.6f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

print("\n💡 图解说明：")
print("  红色: 系统故障率（对数刻度）")
print("  绿色: 系统正常率")
print("  1块硬盘: 故障率 2.89%")
print("  2块硬盘: 故障率 0.08%（降低 36 倍）")
print("  3块硬盘: 故障率 0.002%（降低 1000 倍）")
````

## 7. 综合模拟：掷骰子验证加法和乘法规则

````python
# ========== 步骤 8: 模拟验证加法规则 ==========

# 掷一个公平骰子
# 事件 A: 掷出偶数 {2, 4, 6}
# 事件 B: 掷出大于 4 {5, 6}
# A∪B: {2, 4, 5, 6}
# A∩B: {6}

n_rolls = 100000
rolls = np.random.randint(1, 7, size=n_rolls)

# 理论概率
prob_A_theory = 3/6  # 偶数
prob_B_theory = 2/6  # 大于4
prob_AB_theory = 1/6  # 6（偶数且大于4）
prob_A_or_B_theory = prob_A_theory + prob_B_theory - prob_AB_theory

# 模拟概率
event_A = np.isin(rolls, [2, 4, 6])
event_B = rolls > 4
prob_A_sim = np.mean(event_A)
prob_B_sim = np.mean(event_B)
prob_AB_sim = np.mean(event_A & event_B)
prob_A_or_B_sim = np.mean(event_A | event_B)

print("📊 步骤 8: 模拟验证加法规则 - 掷骰子")
print(f"  事件 A: 掷出偶数 {{2, 4, 6}}")
print(f"  事件 B: 掷出大于 4 {{5, 6}}")
print(f"  A∩B: {{6}}")
print(f"  A∪B: {{2, 4, 5, 6}}")
print(f"\n  {'':>20} {'理论':>10} {'模拟':>10}")
print(f"  {'P(A)':>20} {prob_A_theory:>10.4f} {prob_A_sim:>10.4f}")
print(f"  {'P(B)':>20} {prob_B_theory:>10.4f} {prob_B_sim:>10.4f}")
print(f"  {'P(A∩B)':>20} {prob_AB_theory:>10.4f} {prob_AB_sim:>10.4f}")
print(f"  {'P(A∪B) 加法':>20} {prob_A_or_B_theory:>10.4f} {prob_A_or_B_sim:>10.4f}")
print(f"  {'P(A∪B) 直接':>20} {4/6:>10.4f} {'':>10}")
print(f"\n  ✅ 加法规则验证通过！")
````

````python
# ========== 步骤 9: 模拟验证乘法规则 ==========

# 掷两个骰子
# 事件 A: 第一个骰子为 6
# 事件 B: 第二个骰子为 6
# A∩B: 两个都是 6

n_trials = 100000
die1 = np.random.randint(1, 7, size=n_trials)
die2 = np.random.randint(1, 7, size=n_trials)

# 理论概率（独立事件）
prob_A = 1/6
prob_B = 1/6
prob_AB_theory = prob_A * prob_B  # 独立事件

# 模拟概率
event_A = die1 == 6
event_B = die2 == 6
prob_AB_sim = np.mean(event_A & event_B)

print("📊 步骤 9: 模拟验证乘法规则 - 掷两个骰子")
print(f"  事件 A: 第一个骰子为 6")
print(f"  事件 B: 第二个骰子为 6")
print(f"  A 和 B 是独立事件")
print(f"\n  P(A) = 1/6 = {prob_A:.4f}")
print(f"  P(B) = 1/6 = {prob_B:.4f}")
print(f"  P(A∩B) = P(A) × P(B) = {prob_AB_theory:.4f}")
print(f"  模拟 P(A∩B) = {prob_AB_sim:.4f}")
print(f"  偏差 = {abs(prob_AB_sim - prob_AB_theory):.4f}")
print(f"\n  ✅ 乘法规则验证通过！")
````

````python
# ========== 步骤 10: 树形图 - 乘法原理可视化 ==========

# 书中例题: 两个判断题的树形图
# 第一题: T 或 F (各 0.5)
# 第二题: a, b, c, d, e (各 0.2)

print("📊 步骤 10: 树形图 - 乘法原理")
print(f"  第一题: T 或 F (各 P=0.5)")
print(f"  第二题: a, b, c, d, e (各 P=0.2)")
print(f"\n  样本空间:")

outcomes = []
for q1 in ['T', 'F']:
    for q2 in ['a', 'b', 'c', 'd', 'e']:
        outcomes.append(f"({q1}, {q2})")

print(f"  {outcomes}")
print(f"  总结果数: {len(outcomes)}")
print(f"  每个结果的概率: 0.5 × 0.2 = 0.1")
print(f"  P(两题都正确) = P(T,c) = 0.1")
print(f"\n💡 树形图可视化了乘法原理：第一层分支 × 第二层分支 = 总结果")
````

## 📌 核心概念回顾

### 📌 加法原理 (Addition Rule)
- **公式**: $P(A \cup B) = P(A) + P(B) - P(A \cap B)$
- **互斥事件**: $P(A \cup B) = P(A) + P(B)$（因为 $P(A \cap B) = 0$）
- **含义**: "或" → 加法

### 📌 乘法原理 (Multiplication Rule)
- **公式**: $P(A \cap B) = P(A) \times P(B|A)$
- **独立事件**: $P(A \cap B) = P(A) \times P(B)$（因为 $P(B|A) = P(B)$）
- **含义**: "和" → 乘法

### 📌 独立性 (Independence)
- **定义**: P(B|A) = P(B)，即 A 的发生不影响 B
- **有放回**: 独立事件
- **无放回**: 非独立事件（但 5% 准则可简化）

### 📌 对立事件准则 (Complement Rule)
- **公式**: $P(A) = 1 - P(\bar{A})$
- **应用**: "至少一个" 的概率 = 1 - "一个都没有" 的概率

### 🔗 完整流程
```
判断事件类型 → 互斥? → 加法规则
              → 独立? → 乘法规则
              → 非独立? → 条件概率乘法
    ↓
计算 P(A∪B) 或 P(A∩B) → 验证 → 判断显著性
```

## ❌ 常见误区

### ❌ 误区 1: 非互斥事件直接相加
**✓ 正确理解**: 非互斥事件有重叠部分，必须减去 $P(A \cap B)$，否则会重复计算。

### ❌ 误区 2: 混淆 "或" 与 "和"
**✓ 正确理解**: "或" 对应加法（$\cup$），"和" 对应乘法（$\cap$）。"A 或 B" = A∪B，"A 和 B" = A∩B。

### ❌ 误区 3: 将非独立事件当作独立事件
**✓ 正确理解**: 无放回抽样是非独立事件。只有当样本量 ≤ 总体的 5% 时，才可近似为独立事件。

### ❌ 误区 4: 认为非独立事件一定有因果关系
**✓ 正确理解**: 非独立事件只意味着一个事件影响另一个的概率，不一定是直接原因。例如，共用电源的两个灯泡是非独立的，但一个灯泡不是另一个灯泡坏的原因。

### ❌ 误区 5: 误用乘法原理计算 "至少一个"
**✓ 正确理解**: "至少一个" 的概率应使用对立事件：P(至少一个) = 1 - P(一个都没有)。直接计算 "至少一个" 需要处理多种情况，非常复杂。
