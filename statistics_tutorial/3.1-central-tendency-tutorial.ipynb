## 1. 场景设定：什么是集中趋势？

### 🎯 一个直觉问题

假设你拿到了 10 个学生的考试成绩：

```
55, 62, 68, 72, 75, 78, 80, 82, 85, 99
```

你会用什么数字来代表这组数据的"中心"？

这就是**集中趋势**的问题——用一个代表值来描述数据集的中心位置。

### 📖 书中的定义

> 集中趋势的度量指标是描述数据集中心位置的统计量。最常用的三个指标是：均值（Mean）、中位数（Median）和众数（Mode）。

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import matplotlib.font_manager as fm
_cn_fonts = [f.name for f in fm.fontManager.ttflist if any(kw in f.name for kw in ['Hei', 'Song', 'PingFang', 'Arial Unicode', 'Noto Sans CJK', 'SimHei', 'Microsoft YaHei'])]
plt.rcParams['font.sans-serif'] = _cn_fonts[:3] + ['DejaVu Sans'] if _cn_fonts else ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

np.random.seed(42)
print("✅ 库导入完成")
````

## 2. 从 10 个数据点开始：手算均值、中位数、众数

### 🎯 场景

使用 10 个学生的考试成绩，手算三个集中趋势指标。

````python
# ========== 构造 10 个数据点的微型数据集 ==========
scores = [55, 62, 68, 72, 75, 78, 80, 82, 85, 99]

print("📊 原始数据（10 个学生考试成绩）:")
print(f"  {scores}")
print(f"  数据量: n = {len(scores)}")
````

### 📐 均值 (Mean)

均值是最常用的集中趋势指标，计算公式为：

$$\bar{x} = \frac{\sum x_i}{n} = \frac{x_1 + x_2 + \cdots + x_n}{n}$$

其中：
- $\bar{x}$ = 样本均值
- $\sum x_i$ = 所有数据值的总和
- $n$ = 数据个数

````python
# ========== 步骤 1: 计算均值 ==========
print("📊 步骤 1: 计算均值")
print("─" * 50)

# 显示求和过程
sum_str = " + ".join([str(x) for x in scores])
print(f"  求和: {sum_str}")
print(f"      = {sum(scores)}")
print(f"\n  均值 = 总和 / 数据个数")
print(f"       = {sum(scores)} / {len(scores)}")
print(f"       = {sum(scores) / len(scores):.1f}")

# 手算结果
mean_manual = sum(scores) / len(scores)
print(f"\n  ✅ 手算均值: {mean_manual:.1f} 分")
````

### 📐 中位数 (Median)

中位数是将数据按大小排列后，位于中间位置的值。

**计算方法**：
1. 将数据从小到大排列
2. 如果 n 是奇数，中位数 = 第 (n+1)/2 个值
3. 如果 n 是偶数，中位数 = 中间两个值的均值

````python
# ========== 步骤 2: 计算中位数 ==========
print("📊 步骤 2: 计算中位数")
print("─" * 50)

# 排序
sorted_scores = sorted(scores)
print(f"  排序后: {sorted_scores}")
print(f"  数据个数: n = {len(scores)}（偶数）")

# 偶数个数据：取中间两个的均值
n = len(scores)
mid1 = sorted_scores[n//2 - 1]  # 第 5 个
mid2 = sorted_scores[n//2]      # 第 6 个

print(f"\n  中间位置: 第 {n//2} 个和第 {n//2 + 1} 个")
print(f"  第 {n//2} 个值 = {mid1}")
print(f"  第 {n//2 + 1} 个值 = {mid2}")
print(f"\n  中位数 = ({mid1} + {mid2}) / 2")
print(f"         = {(mid1 + mid2) / 2:.1f}")

median_manual = (mid1 + mid2) / 2
print(f"\n  ✅ 手算中位数: {median_manual:.1f} 分")
````

### 📐 众数 (Mode)

众数是数据集中出现次数最多的值。

- 一个数据集可能没有众数（所有值都只出现一次）
- 一个数据集可能有多个众数（多个值出现次数相同且最多）

````python
# ========== 步骤 3: 计算众数 ==========
print("📊 步骤 3: 计算众数")
print("─" * 50)

# 统计每个值出现的次数
from collections import Counter
freq = Counter(scores)

print(f"  各值出现次数:")
for val, count in sorted(freq.items()):
    marker = " ← 最多" if count == max(freq.values()) else ""
    print(f"    {val}: {count} 次{marker}")

# 找出众数
max_freq = max(freq.values())
modes = [val for val, count in freq.items() if count == max_freq]

if max_freq == 1:
    print(f"\n  💡 所有值都只出现一次，没有众数")
elif len(modes) == 1:
    print(f"\n  ✅ 众数: {modes[0]} 分（出现 {max_freq} 次）")
else:
    print(f"\n  ✅ 多个众数: {modes}（都出现 {max_freq} 次）")
````

````python
# ========== 用 scipy 验证 ==========
mean_scipy = np.mean(scores)
median_scipy = np.median(scores)
mode_scipy = stats.mode(scores, keepdims=True)

print("🔬 scipy 验证:")
print("─" * 50)
print(f"  手算均值 = {mean_manual:.1f}, scipy 均值 = {mean_scipy:.1f}")
print(f"  手算中位数 = {median_manual:.1f}, scipy 中位数 = {median_scipy:.1f}")
print(f"  手算众数 = {modes}, scipy 众数 = {mode_scipy.mode}")
print(f"\n  ✅ 验证通过！")
````

## 4. 不同分布形态下的指标选择

### 📖 书中的指导原则

| 分布形态 | 均值 vs 中位数 | 推荐使用 |
|----------|---------------|----------|
| 对称分布 | 均值 ≈ 中位数 | 均值 |
| 右偏分布 | 均值 > 中位数 | 中位数 |
| 左偏分布 | 均值 < 中位数 | 中位数 |
| 有异常值 | 均值受影响大 | 中位数 |

````python
# ========== 不同分布形态下的均值和中位数 ==========

# 生成三种分布的数据
symmetric = np.random.normal(loc=50, scale=10, size=500)
right_skew = np.random.lognormal(mean=3.5, sigma=0.4, size=500)
left_skew = 100 - np.random.lognormal(mean=3.5, sigma=0.4, size=500)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

datasets = [symmetric, right_skew, left_skew]
titles = ['Symmetric', 'Right-skewed', 'Left-skewed']
colors = ['steelblue', '#e74c3c', '#2ecc71']

for ax, data, title, color in zip(axes, datasets, titles, colors):
    ax.hist(data, bins=20, color=color, alpha=0.7, edgecolor='white', density=True)
    mean_val = np.mean(data)
    median_val = np.median(data)
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.1f}')
    ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median = {median_val:.1f}')
    ax.set_xlabel('Value', fontsize=12)
    ax.set_ylabel('Density', fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n📊 三种分布的均值和中位数:")
print("─" * 50)
for data, name in zip(datasets, titles):
    print(f"  {name:12s}: 均值 = {np.mean(data):.1f}, 中位数 = {np.median(data):.1f}")

print(f"\n💡 解读:")
print(f"  对称分布：均值 ≈ 中位数")
print(f"  右偏分布：均值 > 中位数（被高值拉高）")
print(f"  左偏分布：均值 < 中位数（被低值拉低）")
````

模拟 500 个通勤时间数据，计算集中趋势指标。

## 6. 加权均值

### 📐 加权均值的公式

当不同数据值的重要性不同时，需要使用加权均值：

$$\bar{x}_w = \frac{\sum (w_i \cdot x_i)}{\sum w_i}$$

其中：
- $w_i$ = 第 i 个数据的权重
- $x_i$ = 第 i 个数据的值

### 📖 书中的例子

> 例 8：计算平均绩点（GPA）
>
> 假设一个学生修了五门课：A(3学分)、A(4学分)、B(3学分)、C(3学分)、F(1学分)
>
> 评分标准：A=4, B=3, C=2, D=1, F=0

````python
# ========== 步骤 4: 计算加权均值 ==========
print("📊 步骤 4: 计算加权均值（GPA）")
print("─" * 50)

# 课程数据
grades = ['A', 'A', 'B', 'C', 'F']
credits = [3, 4, 3, 3, 1]
grade_points = {'A': 4, 'B': 3, 'C': 2, 'D': 1, 'F': 0}

# 转换为分数
points = [grade_points[g] for g in grades]

# 计算加权和
weighted_sum = sum(w * x for w, x in zip(credits, points))
total_credits = sum(credits)

print(f"  课程成绩: {grades}")
print(f"  学分数:   {credits}")
print(f"  分数:     {points}")
print(f"\n  计算过程:")
for i in range(len(grades)):
    print(f"    {grades[i]}({credits[i]}学分) = {points[i]} × {credits[i]} = {points[i] * credits[i]}")

print(f"\n  加权和 = {weighted_sum}")
print(f"  总学分 = {total_credits}")
print(f"\n  加权均值 = {weighted_sum} / {total_credits}")
print(f"           = {weighted_sum / total_credits:.2f}")

gpa = weighted_sum / total_credits
print(f"\n  ✅ GPA = {gpa:.2f}")
````

## 7. 截尾均值

### 📐 截尾均值的定义

截尾均值是去掉数据两端一定比例后计算的均值。它可以减少异常值的影响。

例如，10% 截尾均值：去掉最小的 10% 和最大的 10% 后计算均值。

````python
# ========== 步骤 5: 计算截尾均值 ==========
print("📊 步骤 5: 计算截尾均值")
print("─" * 50)

# 使用 10% 截尾均值
trimmed_mean = stats.trim_mean(commute_time, proportiontocut=0.1)

print(f"  原始均值 = {mean_val:.1f} 分钟")
print(f"  10% 截尾均值 = {trimmed_mean:.1f} 分钟")
print(f"  中位数 = {median_val:.1f} 分钟")

print(f"\n💡 解读:")
print(f"  截尾均值介于均值和中位数之间")
print(f"  它比均值更稳健，但比中位数使用了更多信息")
print(f"  在有异常值的情况下，截尾均值是很好的选择")
````

````python
# ========== 可视化：三种均值的比较 ==========
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(commute_time, bins=15, density=True, color='steelblue', alpha=0.7, edgecolor='white')
ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean_val:.1f}')
ax.axvline(trimmed_mean, color='orange', linestyle='--', linewidth=2, 
           label=f'Trimmed Mean (10%) = {trimmed_mean:.1f}')
ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median = {median_val:.1f}')

ax.set_xlabel('Commute Time (min)', fontsize=12)
ax.set_ylabel('Density', fontsize=12)
ax.set_title('Comparison of Mean, Trimmed Mean, and Median', 
             fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明:")
print(f"  红色：均值（受高值影响最大）")
print(f"  橙色：截尾均值（去掉两端 10%）")
print(f"  绿色：中位数（最稳健）")
print(f"  三者的大小关系：均值 > 截尾均值 > 中位数（右偏分布）")
````

## 8. 从频数分布表计算均值

### 📐 公式 3-2

当数据以频数分布表的形式呈现时，均值的计算公式为：

$$\bar{x} = \frac{\sum (f \cdot x_m)}{\sum f}$$

其中：
- $f$ = 每组的频数
- $x_m$ = 每组的组中值

### 📖 书中的例子

> 例 7：根据洛杉矶每日通勤时间的频数分布表计算均值

````python
# ========== 步骤 6: 从频数分布表计算均值 ==========
print("📊 步骤 6: 从频数分布表计算均值")
print("─" * 50)

# 书中的频数分布表
freq_table = pd.DataFrame({
    '组': ['0~14', '15~29', '30~44', '45~59', '60~74', '75~89', '90~104'],
    '频数': [6, 18, 14, 5, 5, 1, 1],
    '组中值': [7.0, 22.0, 37.0, 52.0, 67.0, 82.0, 97.0]
})

# 计算 f * x_m
freq_table['f × xm'] = freq_table['频数'] * freq_table['组中值']

print("  频数分布表:")
print(freq_table.to_string(index=False))

# 计算均值
sum_fx = (freq_table['频数'] * freq_table['组中值']).sum()
sum_f = freq_table['频数'].sum()
mean_from_table = sum_fx / sum_f

print(f"\n  Σ(f × xm) = {sum_fx}")
print(f"  Σf = {sum_f}")
print(f"\n  均值 = {sum_fx} / {sum_f}")
print(f"       = {mean_from_table:.1f} 分钟")

print(f"\n💡 注意:")
print(f"  这是近似值（假设每组所有值等于组中值）")
print(f"  使用原始数据计算的均值 = 31.4 分钟（更准确）")
````

## 📌 核心概念回顾

### 📌 均值 (Mean)
- **公式**: $\bar{x} = \frac{\sum x_i}{n}$
- **含义**: 数据的算术平均
- **特点**: 受异常值影响大，使用所有数据信息
- **适用**: 对称分布、无异常值

### 📌 中位数 (Median)
- **定义**: 排序后位于中间位置的值
- **含义**: 将数据分为两半
- **特点**: 不受异常值影响，最稳健
- **适用**: 偏态分布、有异常值

### 📌 众数 (Mode)
- **定义**: 出现次数最多的值
- **含义**: 最常见的值
- **特点**: 可能没有或有多个
- **适用**: 分类数据、离散数据

### 📌 加权均值 (Weighted Mean)
- **公式**: $\bar{x}_w = \frac{\sum (w_i \cdot x_i)}{\sum w_i}$
- **含义**: 不同数据有不同重要性
- **应用**: GPA 计算、股票指数

### 📌 截尾均值 (Trimmed Mean)
- **定义**: 去掉两端一定比例后计算的均值
- **含义**: 减少异常值影响
- **特点**: 介于均值和中位数之间

### 🔗 完整流程
```
收集数据 → 排序 → 计算均值 → 计算中位数 → 计算众数 → 判断分布形态 → 选择合适的指标
    ↓         ↓        ↓          ↓          ↓           ↓              ↓
  原始数据   升序排列   总和/n    中间位置    最高频     均值vs中位数    报告结果
```

### 📝 集中趋势指标汇总

| 指标 | 公式 | 优点 | 缺点 | 适用场景 |
|------|------|------|------|----------|
| 均值 | Σx/n | 使用所有信息 | 受异常值影响 | 对称分布 |
| 中位数 | 中间值 | 稳健 | 信息利用不充分 | 偏态分布 |
| 众数 | 最高频 | 适用于分类数据 | 可能不存在 | 分类数据 |
| 加权均值 | Σ(wx)/Σw | 考虑重要性 | 需要权重 | GPA、指数 |
| 截尾均值 | 去两端后均值 | 减少异常值影响 | 丢弃部分数据 | 有异常值 |

---

## ❌ 常见误区

### ❌ 误区 1: 均值总是最好的集中趋势指标
**✓ 正确理解**: 均值受异常值影响很大。在偏态分布或有异常值的情况下，中位数或截尾均值可能更合适。

### ❌ 误区 2: 中位数就是中间那个数
**✓ 正确理解**: 只有当 n 是奇数时，中位数才是排序后正中间的那个数。当 n 是偶数时，中位数是中间两个数的均值。

### ❌ 误区 3: 众数一定存在
**✓ 正确理解**: 如果所有值都只出现一次，则没有众数。一个数据集也可能有多个众数（多峰分布）。

### ❌ 误区 4: 计算身份证号码的均值是有意义的
**✓ 正确理解**: 身份证号码是名目数据，只是标签，不能用于计算均值。只有定量数据才能计算均值。

### ❌ 误区 5: 均值一定在数据范围内
**✓ 正确理解**: 均值不一定等于某个实际数据值。例如，3 和 5 的均值是 4，但 4 不在原始数据中。
