## 1. 场景设定：标准正态分布

### 🎯 问题
骨矿物质密度测试有助于确定骨质疏松症的存在或易感性。测试结果通常用 **z 分数**来衡量。

z 分数的总体服从均值为 0，标准差为 1 的正态分布——这就是**标准正态分布**。

### 📖 书中的定义
> 标准正态分布是一个具有以下特性的概率分布：
> 1. 该分布是正态分布，其曲线呈钟形
> 2. 总体均值为 $\mu = 0$
> 3. 总体标准差为 $\sigma = 1$
> 4. 密度曲线下的面积等于 1

### 💡 关键洞察
任何正态分布都可以通过标准化转换为标准正态分布。标准正态分布是统计推断的基石。

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 设置中文字体和样式
import matplotlib.font_manager as fm
_cn_fonts = [f.name for f in fm.fontManager.ttflist if any(kw in f.name for kw in ['Hei', 'Song', 'PingFang', 'Arial Unicode', 'Noto Sans CJK', 'SimHei', 'Microsoft YaHei'])]
plt.rcParams['font.sans-serif'] = _cn_fonts[:3] + ['DejaVu Sans'] if _cn_fonts else ['DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
sns.set_style('whitegrid')

np.random.seed(42)
print("✅ 库导入完成")
````

## 2. 标准正态分布的可视化

### 📐 分布特征
标准正态分布 $N(0, 1)$ 的密度函数：
$$f(z) = \frac{1}{\sqrt{2\pi}} e^{-z^2/2}$$

### 📐 关键性质
- 曲线关于 $z=0$ 对称
- 总面积 = 1
- 约 68% 的面积在 $\pm 1$ 之间
- 约 95% 的面积在 $\pm 1.96$ 之间
- 约 99.7% 的面积在 $\pm 3$ 之间

````python
# ========== 步骤 1: 绘制标准正态分布 ==========

z = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(z, 0, 1)

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制标准正态分布曲线
ax.plot(z, y, 'b-', linewidth=2, label='Standard Normal N(0,1)')
ax.fill_between(z, 0, y, alpha=0.1, color='steelblue')

# 标记关键值
key_values = [-3, -2, -1, 0, 1, 2, 3]
for val in key_values:
    ax.axvline(x=val, color='gray', linestyle=':', alpha=0.5)
    ax.text(val, -0.01, f'{val}', ha='center', fontsize=9)

# 标记 ±1, ±1.96, ±3 的区域
ax.axvline(x=-1.96, color='#e74c3c', linestyle='--', alpha=0.7, linewidth=2)
ax.axvline(x=1.96, color='#e74c3c', linestyle='--', alpha=0.7, linewidth=2)

ax.set_xlabel('z Score', fontsize=12)
ax.set_ylabel('Probability Density', fontsize=12)
ax.set_title('Standard Normal Distribution N(0, 1)', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  蓝色曲线：标准正态分布 N(0, 1)。")
print(f"  红色虚线：z = ±1.96，包含约 95% 的面积。")
print(f"  灰色虚线：z = -3, -2, -1, 0, 1, 2, 3 的位置。")
````

## 3. 给定 z 分数求概率（面积）

### 📐 核心概念
标准正态分布下，z 分数左侧的面积就是累积概率：
$$P(Z < z) = \text{左侧累积面积}$$

### 📖 书中的例子
> 骨密度测试：求 $P(Z < 1.27)$
> 即求 z = 1.27 左侧的面积

````python
# ========== 步骤 2: 求 z = 1.27 左侧的面积 ==========

z_value = 1.27

# 用 scipy 计算左侧累积概率
prob_left = stats.norm.cdf(z_value, 0, 1)

print(f"\n📊 步骤 2: 求 z = {z_value} 左侧的面积")
print(f"  P(Z < {z_value}) = {prob_left:.4f}")
print(f"  即约 {prob_left*100:.2f}% 的数据低于 z = {z_value}")

# 验证：右侧面积
prob_right = 1 - prob_left
print(f"\n  P(Z > {z_value}) = 1 - P(Z < {z_value}) = {prob_right:.4f}")
print(f"  即约 {prob_right*100:.2f}% 的数据高于 z = {z_value}")
````

````python
# ========== 可视化: z = 1.27 左侧的面积 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制标准正态分布
z = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(z, 0, 1)
ax.plot(z, y, 'b-', linewidth=2)

# 填充 z < 1.27 的区域
z_fill = np.linspace(-4, z_value, 500)
y_fill = stats.norm.pdf(z_fill, 0, 1)
ax.fill_between(z_fill, 0, y_fill, alpha=0.3, color='steelblue', 
                label=f'P(Z < {z_value}) = {prob_left:.4f}')

# 标记 z = 1.27
ax.axvline(x=z_value, color='#e74c3c', linestyle='--', linewidth=2)
ax.text(z_value, -0.02, f'z = {z_value}', ha='center', fontsize=11, fontweight='bold')

ax.set_xlabel('z Score', fontsize=12)
ax.set_ylabel('Probability Density', fontsize=12)
ax.set_title(f'Standard Normal: Area to the Left of z = {z_value}', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  阴影区域：z < {z_value} 的面积 = {prob_left:.4f}。")
print(f"  红色虚线：z = {z_value} 的位置。")
print(f"  解读：约 {prob_left*100:.2f}% 的骨密度测试分数低于 1.27。")
````

## 4. 求右侧面积

### 📐 计算公式
右侧面积 = 1 - 左侧面积：
$$P(Z > z) = 1 - P(Z < z)$$

### 📖 书中的例子
> 骨密度测试：求 $P(Z > -1.00)$
> 即求 z = -1.00 右侧的面积

````python
# ========== 步骤 3: 求 z = -1.00 右侧的面积 ==========

z_value_right = -1.00

# 左侧面积
prob_left_neg1 = stats.norm.cdf(z_value_right, 0, 1)
# 右侧面积
prob_right_neg1 = 1 - prob_left_neg1

print(f"\n📊 步骤 3: 求 z = {z_value_right} 右侧的面积")
print(f"  P(Z < {z_value_right}) = {prob_left_neg1:.4f}（左侧面积）")
print(f"  P(Z > {z_value_right}) = 1 - {prob_left_neg1:.4f} = {prob_right_neg1:.4f}（右侧面积）")
print(f"  即约 {prob_right_neg1*100:.2f}% 的骨密度读数高于 -1.00")

# 验证
prob_right_scipy = 1 - stats.norm.cdf(z_value_right, 0, 1)
print(f"\n🔬 验证: 用 scipy 计算 P(Z > {z_value_right}) = {prob_right_scipy:.4f}")
print(f"  ✅ 验证通过！")
````

````python
# ========== 可视化: z = -1.00 右侧的面积 ==========

fig, ax = plt.subplots(figsize=(12, 6))

# 绘制标准正态分布
z = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(z, 0, 1)
ax.plot(z, y, 'b-', linewidth=2)

# 填充 z > -1.00 的区域
z_fill = np.linspace(z_value_right, 4, 500)
y_fill = stats.norm.pdf(z_fill, 0, 1)
ax.fill_between(z_fill, 0, y_fill, alpha=0.3, color='#2ecc71', 
                label=f'P(Z > {z_value_right}) = {prob_right_neg1:.4f}')

# 标记 z = -1.00
ax.axvline(x=z_value_right, color='#e74c3c', linestyle='--', linewidth=2)
ax.text(z_value_right, -0.02, f'z = {z_value_right}', ha='center', fontsize=11, fontweight='bold')

ax.set_xlabel('z Score', fontsize=12)
ax.set_ylabel('Probability Density', fontsize=12)
ax.set_title(f'Standard Normal: Area to the Right of z = {z_value_right}', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  绿色区域：z > {z_value_right} 的面积 = {prob_right_neg1:.4f}。")
print(f"  红色虚线：z = {z_value_right} 的位置。")
print(f"  解读：约 {prob_right_neg1*100:.2f}% 的骨密度水平高于 -1.00（正常范围）。")
````

## 5. 求两个 z 分数之间的面积

### 📐 计算公式
两个 z 分数之间的面积 = 右侧面积 - 左侧面积：
$$P(z_1 < Z < z_2) = P(Z < z_2) - P(Z < z_1)$$

### 📖 书中的例子
> 骨密度测试：求 $P(-2.50 < Z < -1.00)$
> 骨密度读数在 -1.00 和 -2.50 之间表示骨质减少症

````python
# ========== 步骤 4: 求两个 z 分数之间的面积 ==========

z_low = -2.50
z_high = -1.00

# 左侧面积
prob_low = stats.norm.cdf(z_low, 0, 1)
prob_high = stats.norm.cdf(z_high, 0, 1)

# 之间的面积
prob_between = prob_high - prob_low

print(f"\n📊 步骤 4: 求 z = {z_low} 和 z = {z_high} 之间的面积")
print(f"  P(Z < {z_low}) = {prob_low:.4f}")
print(f"  P(Z < {z_high}) = {prob_high:.4f}")
print(f"\n  P({z_low} < Z < {z_high}) = P(Z < {z_high}) - P(Z < {z_low})")
print(f"  P({z_low} < Z < {z_high}) = {prob_high:.4f} - {prob_low:.4f}")
print(f"  P({z_low} < Z < {z_high}) = {prob_between:.4f}")
print(f"  即约 {prob_between*100:.2f}% 的人患有骨质减少症")
````

````python
# ========== 可视化: 两个 z 分数之间的面积 ==========

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

z = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(z, 0, 1)

# 图1: z < -1.00 的面积
axes[0].plot(z, y, 'b-', linewidth=2)
z_fill1 = np.linspace(-4, z_high, 500)
y_fill1 = stats.norm.pdf(z_fill1, 0, 1)
axes[0].fill_between(z_fill1, 0, y_fill1, alpha=0.3, color='steelblue')
axes[0].axvline(x=z_high, color='#e74c3c', linestyle='--', linewidth=2)
axes[0].set_title(f'P(Z < {z_high}) = {prob_high:.4f}', fontsize=12, fontweight='bold')
axes[0].set_xlabel('z Score', fontsize=10)
axes[0].set_ylabel('Density', fontsize=10)
axes[0].grid(alpha=0.3)

# 图2: z < -2.50 的面积
axes[1].plot(z, y, 'b-', linewidth=2)
z_fill2 = np.linspace(-4, z_low, 500)
y_fill2 = stats.norm.pdf(z_fill2, 0, 1)
axes[1].fill_between(z_fill2, 0, y_fill2, alpha=0.3, color='steelblue')
axes[1].axvline(x=z_low, color='#e74c3c', linestyle='--', linewidth=2)
axes[1].set_title(f'P(Z < {z_low}) = {prob_low:.4f}', fontsize=12, fontweight='bold')
axes[1].set_xlabel('z Score', fontsize=10)
axes[1].set_ylabel('Density', fontsize=10)
axes[1].grid(alpha=0.3)

# 图3: 之间的面积
axes[2].plot(z, y, 'b-', linewidth=2)
z_fill3 = np.linspace(z_low, z_high, 500)
y_fill3 = stats.norm.pdf(z_fill3, 0, 1)
axes[2].fill_between(z_fill3, 0, y_fill3, alpha=0.3, color='#2ecc71',
                     label=f'P({z_low} < Z < {z_high}) = {prob_between:.4f}')
axes[2].axvline(x=z_low, color='#e74c3c', linestyle='--', linewidth=2)
axes[2].axvline(x=z_high, color='#e74c3c', linestyle='--', linewidth=2)
axes[2].set_title(f'Area Between z = {z_low} and {z_high}', fontsize=12, fontweight='bold')
axes[2].set_xlabel('z Score', fontsize=10)
axes[2].set_ylabel('Density', fontsize=10)
axes[2].legend(fontsize=9)
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  左图：z < -1.00 的面积。")
print(f"  中图：z < -2.50 的面积。")
print(f"  右图：z 在 -2.50 和 -1.00 之间的面积 = {prob_between:.4f}。")
print(f"  计算方法：右图面积 = 中图面积 - 左图面积。")
````

## 6. 给定面积求 z 分数（临界值）

### 📐 核心概念
有时候我们需要反过来求：已知概率（面积），求对应的 z 分数。

### 📐 临界值的定义
对于标准正态分布，**临界值**是一个 z 分数，它位于分开显著低或显著高的分数的分界线上。

### 📐 常用临界值
| 临界值 | 左侧面积 | 右侧面积 | 含义 |
|--------|----------|----------|------|
| $z = \pm 1.645$ | 0.9500 / 0.0500 | 0.0500 / 0.9500 | 90% 置信区间 |
| $z = \pm 1.96$ | 0.9750 / 0.0250 | 0.0250 / 0.9750 | 95% 置信区间 |
| $z = \pm 2.575$ | 0.9950 / 0.0050 | 0.0050 / 0.9950 | 99% 置信区间 |

````python
# ========== 步骤 5: 求给定面积对应的 z 分数 ==========

# 场景1: 求第 95 百分位数对应的 z 分数
# 即 P(Z < z) = 0.95
prob_95 = 0.95
z_95 = stats.norm.ppf(prob_95, 0, 1)

print(f"\n📊 步骤 5: 求给定面积对应的 z 分数")
print(f"\n  场景1: 求第 95 百分位数")
print(f"  P(Z < z) = {prob_95}")
print(f"  z = {z_95:.4f}")
print(f"  → 95% 的骨密度分数小于或等于 {z_95:.4f}")

# 场景2: 求分开最低 2.5% 和最高 2.5% 的 z 分数
# 即 P(Z < z) = 0.025 和 P(Z < z) = 0.975
prob_low_25 = 0.025
prob_high_25 = 0.975
z_low_25 = stats.norm.ppf(prob_low_25, 0, 1)
z_high_25 = stats.norm.ppf(prob_high_25, 0, 1)

print(f"\n  场景2: 求分开最低 2.5% 和最高 2.5% 的 z 分数")
print(f"  P(Z < z) = {prob_low_25} → z = {z_low_25:.4f}")
print(f"  P(Z < z) = {prob_high_25} → z = {z_high_25:.4f}")
print(f"  → 95% 的骨密度分数在 {z_low_25:.4f} 和 {z_high_25:.4f} 之间")

# 验证
print(f"\n🔬 验证:")
print(f"  P(Z < {z_low_25:.4f}) = {stats.norm.cdf(z_low_25, 0, 1):.4f}")
print(f"  P(Z < {z_high_25:.4f}) = {stats.norm.cdf(z_high_25, 0, 1):.4f}")
print(f"  ✅ 验证通过！")
````

````python
# ========== 可视化: 临界值 ==========

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

z = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(z, 0, 1)

# 图1: 第 95 百分位数
axes[0].plot(z, y, 'b-', linewidth=2)
z_fill1 = np.linspace(-4, z_95, 500)
y_fill1 = stats.norm.pdf(z_fill1, 0, 1)
axes[0].fill_between(z_fill1, 0, y_fill1, alpha=0.3, color='steelblue',
                     label=f'P(Z < {z_95:.3f}) = 0.95')
axes[0].axvline(x=z_95, color='#e74c3c', linestyle='--', linewidth=2)
axes[0].set_title('95th Percentile', fontsize=12, fontweight='bold')
axes[0].set_xlabel('z Score', fontsize=10)
axes[0].set_ylabel('Density', fontsize=10)
axes[0].legend(fontsize=10)
axes[0].grid(alpha=0.3)

# 图2: 分开最低和最高 2.5%
axes[1].plot(z, y, 'b-', linewidth=2)
# 中间 95% 的区域
z_fill_mid = np.linspace(z_low_25, z_high_25, 500)
y_fill_mid = stats.norm.pdf(z_fill_mid, 0, 1)
axes[1].fill_between(z_fill_mid, 0, y_fill_mid, alpha=0.3, color='#2ecc71',
                     label=f'95% area')
# 两侧 2.5% 的区域
z_fill_left = np.linspace(-4, z_low_25, 500)
y_fill_left = stats.norm.pdf(z_fill_left, 0, 1)
axes[1].fill_between(z_fill_left, 0, y_fill_left, alpha=0.3, color='#e74c3c')
z_fill_right = np.linspace(z_high_25, 4, 500)
y_fill_right = stats.norm.pdf(z_fill_right, 0, 1)
axes[1].fill_between(z_fill_right, 0, y_fill_right, alpha=0.3, color='#e74c3c')

axes[1].axvline(x=z_low_25, color='#e74c3c', linestyle='--', linewidth=2)
axes[1].axvline(x=z_high_25, color='#e74c3c', linestyle='--', linewidth=2)
axes[1].set_title('Critical Values: z = ±1.96', fontsize=12, fontweight='bold')
axes[1].set_xlabel('z Score', fontsize=10)
axes[1].set_ylabel('Density', fontsize=10)
axes[1].legend(fontsize=10)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n💡 图解说明：")
print(f"  左图：第 95 百分位数对应的 z = {z_95:.3f}。")
print(f"  右图：z = ±1.96 将分布分为三部分：")
print(f"    - 左侧 2.5%（红色）：z < -1.96")
print(f"    - 中间 95%（绿色）：-1.96 < z < 1.96")
print(f"    - 右侧 2.5%（红色）：z > 1.96")
````

## 7. 临界值的数学符号

### 📐 符号说明
$z_\alpha$ 表示**右侧面积为 $\alpha$** 的 z 分数。

### 📐 常用符号
- $z_{0.025} = 1.96$：右侧面积为 0.025 的 z 分数
- $z_{0.05} = 1.645$：右侧面积为 0.05 的 z 分数
- $z_{0.005} = 2.575$：右侧面积为 0.005 的 z 分数

### 💡 注意
$z_\alpha$ 中的 $\alpha$ 是**右侧**的面积，但 scipy 给出的是左侧累积面积，所以需要用 $1 - \alpha$ 来计算。

````python
# ========== 步骤 6: 临界值的数学符号 ==========

# 求 z_{0.025}
alpha_025 = 0.025
# z_{0.025} 右侧面积为 0.025，左侧面积为 0.975
z_025 = stats.norm.ppf(1 - alpha_025, 0, 1)

print(f"\n📊 步骤 6: 临界值的数学符号")
print(f"\n  求 z_{{{alpha_025}}}：")
print(f"  z_{{{alpha_025}}} 表示右侧面积为 {alpha_025} 的 z 分数")
print(f"  左侧面积 = 1 - {alpha_025} = {1 - alpha_025}")
print(f"  z_{{{alpha_025}}} = {z_025:.4f}")

# 验证
print(f"\n🔬 验证:")
print(f"  P(Z < {z_025:.4f}) = {stats.norm.cdf(z_025, 0, 1):.4f}")
print(f"  P(Z > {z_025:.4f}) = {1 - stats.norm.cdf(z_025, 0, 1):.4f}")
print(f"  ✅ 验证通过！")

# 常用临界值汇总
print(f"\n📝 常用临界值汇总:")
print(f"  {'α':<10} {'z_α':<12} {'左侧面积':<12}")
print(f"  {'-'*35}")
alphas = [0.10, 0.05, 0.025, 0.01, 0.005]
for a in alphas:
    z_a = stats.norm.ppf(1 - a, 0, 1)
    print(f"  {a:<10.3f} {z_a:<12.4f} {1-a:<12.4f}")
````

## 8. 模拟验证

### 📐 模拟实验
通过模拟实验，我们可以验证标准正态分布的理论概率。

### 📖 实验设计
> 生成 100000 个标准正态分布的随机数，
> 统计各种 z 分数范围内的比例，并与理论值比较。

````python
# ========== 步骤 7: 模拟验证 ==========

# 生成模拟数据
n_simulations = 100000
simulated_z = np.random.normal(0, 1, n_simulations)

# 计算各种比例
prop_below_127 = np.mean(simulated_z < 1.27)
prop_above_neg1 = np.mean(simulated_z > -1.00)
prop_between = np.mean((simulated_z > -2.50) & (simulated_z < -1.00))
prop_within_196 = np.mean(np.abs(simulated_z) < 1.96)

# 理论值
theory_below_127 = stats.norm.cdf(1.27, 0, 1)
theory_above_neg1 = 1 - stats.norm.cdf(-1.00, 0, 1)
theory_between = stats.norm.cdf(-1.00, 0, 1) - stats.norm.cdf(-2.50, 0, 1)
theory_within_196 = stats.norm.cdf(1.96, 0, 1) - stats.norm.cdf(-1.96, 0, 1)

print(f"\n📊 步骤 7: 模拟验证")
print(f"  模拟次数: {n_simulations}")
print(f"\n  {'指标':<30} {'模拟值':<12} {'理论值':<12} {'差异':<10}")
print(f"  {'-'*65}")
print(f"  {'P(Z < 1.27)':<30} {prop_below_127:<12.4f} {theory_below_127:<12.4f} {abs(prop_below_127-theory_below_127):<10.4f}")
print(f"  {'P(Z > -1.00)':<30} {prop_above_neg1:<12.4f} {theory_above_neg1:<12.4f} {abs(prop_above_neg1-theory_above_neg1):<10.4f}")
print(f"  {'P(-2.50 < Z < -1.00)':<30} {prop_between:<12.4f} {theory_between:<12.4f} {abs(prop_between-theory_between):<10.4f}")
print(f"  {'P(|Z| < 1.96)':<30} {prop_within_196:<12.4f} {theory_within_196:<12.4f} {abs(prop_within_196-theory_within_196):<10.4f}")

print(f"\n💡 模拟结果与理论值非常接近，验证了标准正态分布的正确性。")
````

## 📌 核心概念回顾

### 📌 标准正态分布 (Standard Normal Distribution)
- **定义**: 均值为 0、标准差为 1 的正态分布
- **符号**: $N(0, 1)$
- **特征**: 钟形曲线，关于 $z=0$ 对称，总面积 = 1

### 📌 z 分数
- **定义**: $z = \frac{x - \mu}{\sigma}$，衡量数据点偏离均值的标准差个数
- **含义**: z 分数是沿横坐标轴的距离
- **用途**: 比较不同量纲的数据

### 📌 概率计算
- **给定 z 求面积**: $P(Z < z) = \text{cdf}(z)$
- **给定面积求 z**: $z = \text{ppf}(\text{面积})$
- **临界值**: $z_\alpha$ 表示右侧面积为 $\alpha$ 的 z 分数

### 📌 经验法则
- 约 68% 的数据在 $\mu \pm 1\sigma$ 之间
- 约 95% 的数据在 $\mu \pm 1.96\sigma$ 之间
- 约 99.7% 的数据在 $\mu \pm 3\sigma$ 之间

### 🔗 完整流程
```
确定问题 → 画图标注 → 计算概率/临界值 → 解读结果
    ↓           ↓           ↓               ↓
  求面积?    标记区域    cdf 或 ppf      实际意义
  求 z 值?   阴影标注    scipy 验证      统计推断
```

## ❌ 常见误区

### ❌ 误区 1: 混淆 z 分数和面积
**✓ 正确理解**: z 分数是横坐标轴上的距离，面积（概率）是曲线下的区域。不要把 z 分数当作概率使用。

### ❌ 误区 2: 忘记标准正态分布只适用于 μ=0, σ=1
**✓ 正确做法**: 标准正态分布只适用于均值为 0、标准差为 1 的情况。对于非标准正态分布，需要先标准化（下一节内容）。

### ❌ 误区 3: 混淆 $z_\alpha$ 中的 $\alpha$ 位置
**✓ 正确理解**: $z_\alpha$ 中的 $\alpha$ 是**右侧**面积，但 scipy 给出的是左侧累积面积，所以需要用 $1 - \alpha$ 来计算。

### ❌ 误区 4: 忘记面积必须在 0 和 1 之间
**✓ 正确理解**: 概率（面积）的取值范围是 $[0, 1]$，不可能大于 1 或小于 0。

### ❌ 误区 5: 把 z = 0 当作"没有数据"
**✓ 正确理解**: z = 0 表示数据等于均值，不是"没有数据"。z 分数衡量的是偏离均值的程度。
