### 哨兵值 (Sentinel Values) => NA, NULL

缺失值： NAN，NULL

CSV 文件本质上只是纯文本。在文本中，我们无法直接存储"空"这个概念（因为空字符串 "" 也是一种值）。所以，我们需要约定一些特殊的字符串来代表"空"。这些特殊的字符串被称为哨兵值 (Sentinel Values)。

```python
!cat /Users/d0m999/Desktop/_bot/data_analysis/pydata-book-3rd-edition/examples/ex5.csv
```

```python
result = pd.read_csv('/Users/d0m999/Desktop/_bot/data_analysis/pydata-book-3rd-edition/examples/ex5.csv')
result
```

```python
result.isna()
```

```python
# na_values 参数可以输入多种类型的值来定义缺失值标识符：
# na_values = [999, -1, 'NaN']
# 
# 常见默认缺失值标识符:
#      - 'NaN', 'nan', 'NA', 'N/A'
#      - 'null', 'NULL', 'None'
#      - '' (空字符串)
#      - 'missing', 'MISSING'
#      - '#N/A', '#N/A N/A', '#NA'
```

```python
result = pd.read_csv('/Users/d0m999/Desktop/_bot/data_analysis/pydata-book-3rd-edition/examples/ex5.csv', 
                     na_values={'c':[3, 'NULL']}) 
                     # 定义 3 和 'NULL' 为缺失值
                     # Pandas 内置了一个默认缺失值列表，
                     # 其中已经包含了 'NULL', 'NaN', 'NA', '#N/A' 等常见写法。
result
```

### 分段读取大文件迭代器

要分段读取文件，请将 chunksize 指定为行数：

chunk : 块

```python
chunker = pd.read_csv('/Users/d0m999/Desktop/_bot/data_analysis/pydata-book-3rd-edition/examples/ex6.csv', 
                        chunksize=1000)
chunker
```

```python
tot = pd.Series([], dtype='float64')
for piece in chunker:
    # 从 chunker 中逐块读取 Series
    # 因为 chunker 是TextFileReader 迭代器，所以不能直接应用 dataframe.apply() 方法
    tot = tot.add(piece['key'].value_counts(), fill_value=0)
tot = tot.sort_values(ascending=False)
tot[:10]
```

tot.add(piece['key'].value_counts(), fill_value=0) 语法解析

```python
# 1. 初始化一个空的"记分板"
tot = pd.Series([], dtype='float64')

# 2. 依次处理每一块
for piece in chunker:
    # piece 是一个 DataFrame (例如 1000 行)
    # piece['key'] 是这一块里的 'key' 列 (Series)
    # .value_counts() 统计这一块里每个 key 出现的次数
    
    # 3. 把"这一块的统计"累加到"总记分板"上
    #    fill_value=0 确保新出现的 key 不会变成 NaN
    tot = tot.add(piece['key'].value_counts(), fill_value=0)

# 4. 所有块处理完毕，tot 就是完整的全局统计
tot = tot.sort_values(ascending=False)

   [块1] key: A, B, A    --> value_counts: {A:2, B:1}
                         |
                         v
                    tot = {A:2, B:1}

   [块2] key: B, B, C    --> value_counts: {B:2, C:1}
                         |
                    tot.add(...)
                         |
                         v
                    tot = {A:2, B:3, C:1}

   [块3] key: A, C, C    --> value_counts: {A:1, C:2}
                         |
                    tot.add(...)
                         |
                         v
                    tot = {A:3, B:3, C:3}  <-- ✅ 最终结果
```

### 自定义 csv 格式

定义 csv.Dialect 类必须根据目标文件的实际格式来定义，以确保正确解析数据。

```python
class my_dialect(csv.Dialect):
    lineterminator = '\n' # 行结束符
    delimiter = ',' # 分隔符
    quotechar = '"' # 引号字符
    quoting = csv.QUOTE_MINIMAL 
    # 引号策略: 最小化引号（只对需要引号的字段加引号）, 其他参数: QUOTE_ALL（所有字段都加引号）, QUOTE_NONNUMERIC（非数字字段加引号）, QUOTE_NONE（不加引号）

with open('/Users/d0m999/Desktop/_bot/data_analysis/pydata-book-3rd-edition/examples/ex7.csv') as f:
    reader = list(csv.reader(f, dialect=my_dialect))
reader
```

### Parsing XML with lxml.objectify

```python
from lxml import objectify

path = 'pydata-book-3rd-edition/datasets/mta_perf/Performance_MNR.xml'

with open(path) as f:
    parsed = objectify.parse(f)

root = parsed.getroot() # 获取 XML 文档的根元素
```

```python
data = []
skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ',
               'DESIRED_CHANGE', 'DECIMAL_PLACES']

for elt in root.INDICATOR: # root.INDICATOR 是 lxml.objectify 的语法，返回所有 INDICATOR 元素的迭代器 XML
    el_data = {}
    for child in elt.getchildren(): # getchildren() 返回子元素列表，child 是每个子元素对象
        if child.tag in skip_fields: # child.tag：XML 元素标签名（字符串）
            continue
        el_data[child.tag] = child.pyval # child.pyval：lxml.objectify 自动转换的 Python value（会自动识别数据类型：字符串、整数、浮点数等）
    data.append(el_data)
```

```python
data[:2]
```

```python
perf = pd.DataFrame(data)
perf.head()
```
