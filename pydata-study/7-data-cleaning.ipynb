## 使用函数或映射mapping转换数据

```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

data
```

### map修改值

```python
meat_to_animal = {
  "bacon": "pig",
  "pulled pork": "pig",
  "pastrami": "cow",
  "corned beef": "cow",
  "honey ham": "pig",
  "nova lox": "salmon"
}

data['animal'] = data['food'].map(meat_to_animal) # 修改值

data
```

### map映射值

```python
def get_animal(x):
    return meat_to_animal[x]

get_animal('bacon')
```

```python
data['food'].map(get_animal)
```

```python
data['food'].map(lambda x: meat_to_animal[x]) # lambda函数
```

## 替换值

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

```python
print(data.replace(-999, np.nan))
print('-' * 10)
print(data.replace([-999, -1000], np.nan))
print('-' * 10)
print(data.replace([-999, -1000], [np.nan, 0])) # 分别替换两个不同值
```

```python
print(data.replace({-999: np.nan, -1_000: 0})) # 用 dict-like 语句传递参数
print('-' * 10)
print(f'data 值不变\n{data}\n.replace() 创建副本,不修改原视图')
```

## 重命名轴标签

```python
data = pd.DataFrame(np.arange(12).reshape((3,4)),
                    index=["Ohio", "Colorado", "New York"],
                    columns=["one", "two", "three", "four"])
data
```

```python
data.index.map(lambda x: x.upper())
```

```python
def upper_func(x):
    return x.upper()

data.index.map(upper_func)
```

```python
print(data)
print('data 原视图未改变')
print('-'*10)
data.index = data.index.map(upper_func)
print(data)
print('data.index 被修改')
```

上述方法直接改变了 `data.index`

如果不想改变原视图就实现 rename，可以使用 `rename` 函数

```python
data_rename = data.rename(index=str.title, columns=str.upper) # copy

data_rename
```

`rename` 省去了为 `index` `columns` 赋值的麻烦

```python
data # with modifying the original 
```

## 离散化与分箱 Discretization and Binning

连续数据通常会被离散化或以其他方式分成“箱 bins”进行分析。

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)

age_categories
```

```python
age_categories.codes
```

```python
age_categories.categories
```

```python
age_categories.value_counts()
```

在区间的字符串表示中，括号表示边是开的 （排他的），而方括号表示是闭的 （包含的）。

你可以通过传递 `right =False` 来改变哪一侧是闭合的：

```python
pd.cut(ages, bins, right=False)
```

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]

pd.cut(ages, bins, labels=group_names)
```

如果你把整数个 bin 传递给 `pandas.cut` 而不是显式 bin 的边，它会根据数据中的最小值和最大值计算出等长 bin。

考虑一些均匀分布的数据被切成四分之一的情况：

```python
data = rng.uniform(size=20)
pd.cut(data, 4, precision=2) # 小数点显示精度，只影响显示，不影响计算
# 第一个bin的下界是数据的最小值，最后一个 bin 的上界是数据的最大值
```
