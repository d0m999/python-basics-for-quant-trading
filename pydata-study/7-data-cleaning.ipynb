## 离散化与分箱 Discretization and Binning

连续数据通常会被离散化或以其他方式分成“箱 bins”进行分析。

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)

age_categories
```

```python
age_categories.codes
```

```python
age_categories.value_counts()
```

在区间的字符串表示中，括号表示边是开的 （排他的），而方括号表示是闭的 （包含的）。

你可以通过传递 `right =False` 来改变哪一侧是闭合的：

```python
pd.cut(ages, bins, right=False)
```

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]

pd.cut(ages, bins, labels=group_names)
```

如果你把整数个 bin 传递给 `pandas.cut` 而不是显式 bin 的边，它会根据数据中的最小值和最大值计算出等长 bin。

考虑一些均匀分布的数据被切成四分之一的情况：

```python
data = rng.uniform(size=20)
data_cate = pd.cut(data, 4, precision=2) # 小数点显示精度，只影响显示，不影响计算
# 第一个bin的下界是数据的最小值，最后一个 bin 的上界是数据的最大值
```

```python
data_cate.value_counts()
```

`pandas.qcut` 基于样本分位数对数据进行分组处理。根据数据分布不同，使用 `pandas.cut` 通常不会让每个箱的数据点数相同。由于 `pandas.qcut` 使用样本分位数，你会得到大小大致相同的箱：

```python
data = rng.standard_normal(1000)
quartiles = pd.qcut(data, 4, precision=2)

quartiles
```

```python
quartiles.value_counts()
```

```python
#  可视化分箱过程
#   1 数据点分布:   1    2    3    4                      100   101   102
#   2              |    |    |    |-------- 巨大间隔 -----|    |    |
#   3 
#   4 pd.cut (等宽): |-------箱1-------|-------箱2-------|-------箱3-------|
#   5               宽度相等但点数不均   空箱              点数集中
#   6 
#   7 pd.qcut (等频): |-箱1-| |-箱2-| |-----------箱3------------|
#   8                 ≈2个点  ≈2个点            ≈3个点
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]).value_counts()
```

## 指示/虚拟变量 Computing Indicator/Dummy Variables

  核心思想: <br>将分类变量（categorical variable）
  转换为二进制（0/1）数值矩阵，使分类数据能够被数学模型（如回归、机器学习）处理。

  错误做法（序数编码）：  
   1 # 将分类直接映射为1,2,3...会导致模型误解  
   `2 df['city'] = df['city'].map({'北京': 1, '上海': 2, '广州': 3})`  
   3 # 问题：模型会认为"广州 > 上海 > 北京"（数值大小关系），但这毫无意义！

  正确做法（虚拟变量）：  
   1 # 创建独立的二进制列  
   `2 city_dummies = pd.get_dummies(df['city'], prefix='city')`  
   3 # 结果：city_北京, city_上海, city_广州 三列，每列值为0或1

```python
df['A'] =  df['A'].astype('Int64') # pandas type
df['B'] =  df["B"].astype("string")
df["C"] = df["C"].astype("boolean")
df
```

```python
# fillna
df = df.ffill()

df['A'] =  df['A'].astype('int64') # numpy type: 'int64' or np.int64
df['B'] =  df["B"].astype('str')
df["C"] = df["C"].astype('bool')
print(df)
print('-'*10)
print(df['C'])
```

### Series.str 访问器

  Series 访问器设计的核心智慧：  

   1. 命名空间隔离：避免方法名冲突，`.str`专门处理字符串，`.dt`专门处理时间  
   2. 类型安全：只有合适的数据类型才有对应的访问器，避免无意义的操作  
   3. 语义清晰：`series.contains()`（元素包含）vs `series.str.contains()`（字符串包含）  
   4. 性能优化：访问器方法都是向量化实现，比Python循环快10-100倍  
   5. 链式操作：支持流畅的方法链，代码更简洁优雅  

```python
#     1 Series 对象
#     2 ├── 数据: ['abc123', 'def456', 'ghi789']
#     3 ├── 索引: [0, 1, 2]
#     4 ├── 属性和方法
#     5 │   ├── .values → NumPy数组
#     6 │   ├── .index → 索引对象
#     7 │   ├── .dtype → 数据类型
#     8 │   └── .str → StringMethods访问器 ← 关键！
#     9 │           ├── .findall()
#    10 │           ├── .extract()
#    11 │           ├── .contains()
#    12 │           ├── .replace()
#    13 │           └── ... 更多字符串方法
#    14 └── 其他访问器
#    15     ├── .dt → DateTimeMethods (时间序列)
#    16     ├── .cat → CategoricalMethods (分类数据)
#    17     └── .sparse → SparseMethods (稀疏数据)
```

  为了省内存和算得快，Pandas 搞了个“代号系统”，把重复的字符串变成了整数 ID。这个系统就叫 Categorical。  
  我把“Background and Motivation”这一部分拆解为三个步骤来解析：

  1. 痛点：重复数据的存储浪费

  想象一下，你有一张包含 1000 万行数据的用户表，其中有一列是 `country`（国家）。

  虽然有 1000 万行，但实际上“国家”这个值的种类是非常有限的（大概也就 200 个左右）。这就意味着，像 `"United
  States of America"` 这样很长的字符串，会在你的内存里重复出现几百万次。

  这就好比：
  你在填 1000 张表格。每一张表格都要手写一遍“美利坚合众国”。
   * 不仅手累（占用内存大）
   * 处理起来也慢（比较字符串比比较数字慢得多）。

```python
values = pd.Series(['apple', 'orange', 'apple', 'apple'] * 2)
values
```

```python
values.unique()
```

```python
values.value_counts()
```

  2. 解决方案：维度表（Dimension Table）

  Wes McKinney 在书中提到了 `"Dimension Table"`（维度表），这其实是数据库里的一种经典优化手段。

  我们可以把数据拆成两部分来存：

   1. 词典（也就是书里说的 `Dimension Table`）：只存那 200 个不重复的国家名字。
   2. 编码（Codes）：在原始的 1000 万行数据里，不再存字符串，而是存这 200 个国家对应的整数 ID。

   我为你画了一个 ASCII 图来演示这个转换过程（First Principle: Dictionary-encoded representation）：

```python

#     1 【原始数据】(内存占用巨大)
#     2 +-------+--------------------------+
#     3 |  ID   |         Country          |
#     4 +-------+--------------------------+
#     5 |   1   | United States of America |  <-- 存了24个字符
#     6 |   2   | United States of America |  <-- 又存了24个字符
#     7 |   3   | Japan                    |
#     8 |   4   | United States of America |
#     9 +-------+--------------------------+
#    10 
#    11         ⬇️ 转换 (Encoding) ⬇️
#    12 
#    13 【Pandas Categorical 内部结构】(内存极小)
#    14 
#    15 (1) 词典 (Categories)          (2) 编码 (Codes)
#    16     这里只存唯一值                  这里只存整数(0,1,2...)
#    17 +----+--------------------------+    +-------+--------+
#    18 | ID |  Value (Category)        |    | RowID |  Code  |
#    19 +----+--------------------------+    +-------+--------+
#    20 | 0  | United States of America |    |   1   |   0    | <-- 0 代表 US
#    21 | 1  | Japan                    |    |   2   |   0    |
#    22 +----+--------------------------+    |   3   |   1    | <-- 1 代表 Japan
#    23                                      |   4   |   0    |
#    24                                      +-------+--------+
```

  书中的核心术语解析：
   * Categorical Representation: 就是上面这种“用整数代表字符串”的存储方式。
   * Categories (类别): 那个只存唯一值的词典（如 US, Japan）。
   * Codes (编码): 那个全是整数的数组（如 0, 0, 1, 0）。

```python
values = pd.Series([0, 1, 0, 0] * 2)
dimension_table = pd.Series(['apple', 'orange'])
print(values)
print(dimension_table)
```

```python
dimension_table.take(values)

# .take 表达的意思是：“我有一堆索引，请帮我瞬间映射回真实值”。
#  常见陷阱与注意事项

#    1. 越界错误: 如果你的 dimension_table 只有 3 个元素（索引 0, 1, 2），但你的 values 里包含了一个 5，那么
#       .take() 会抛出 IndexError。
#    2. 处理缺失值: 在真实的 Categorical 数据内部，如果某个值是缺失的（NaN），它的 code 通常会被设为 -1。但在
#       .take() 语法中，不支持 -1，你需要先处理掉这些值。
```

  3. 动机 (Motivation)：我们图什么？

  书中列出了这样做的根本动机，总结起来就是三点：

  A. 内存优化 (Memory)
  存一个整数（比如 `0`）只需要 1 到 4 个字节。存一个字符串 `"United States of America"`
  需要几十个字节。当数据量上亿时，这个压缩效果是惊人的。

  B. 性能提升 (Performance)
  计算机处理数字比处理文字快得多。
   * 排序：直接比大小 `0 < 1` 极快。
   * 分组 (GroupBy)：按整数 ID 分组比按字符串哈希分组更快。

  C. 逻辑顺序 (Semantics) - 这一点常被初学者忽略
  字符串只能按字母排序（Alphabetical），但有时我们需要逻辑排序（Logical）。
   * 字符串排序: `"High" < "Low" < "Medium"` (按首字母 H < L < M，这是错的逻辑)
   * Categorical 排序: 我们可以强制规定顺序：`Low (0) < Medium (1) < High (2)`。

## Categorical Methods

Series containing `categorical` data have several special methods similar to the `Series.str`

```python
#     1 Series 对象
#     2 ├── 数据: ['abc123', 'def456', 'ghi789']
#     3 ├── 索引: [0, 1, 2]
#     4 ├── 属性和方法
#     5 │   ├── .values → NumPy数组
#     6 │   ├── .index → 索引对象
#     7 │   ├── .dtype → 数据类型
#     8 │   └── .str → StringMethods访问器 ← 关键！
#     9 │           ├── .findall()
#    10 │           ├── .extract()
#    11 │           ├── .contains()
#    12 │           ├── .replace()
#    13 │           └── ... 更多字符串方法
#    14 └── 其他访问器
#    15     ├── .dt → DateTimeMethods (时间序列)
#    16     ├── .cat → CategoricalMethods (分类数据)
#    17     └── .sparse → SparseMethods (稀疏数据)
```

```python
s = pd.Series(['a', 'b', 'c', 'd'] * 2)
cat_s = s.astype('category')
cat_s
```

```python
cat_s.cat.categories # .cat 访问器
```

```python
cat_s.cat.codes
```

假设我们知道该数据的实际类别集合超出了数据中观察到的四个值。我们可以用 `set_categories` 方法来更改它们：

```python
actual_categories = ['a', 'b', 'c', 'd', 'e']

cat_s2 = cat_s.cat.set_categories(actual_categories)
cat_s2 # 类别 'e' 虽然未出现，但被包含在 categories 中
```

```python
print(cat_s.value_counts())
print(cat_s2.value_counts()) # 包含 'e' 类别，计数为0
```

在大型数据集中，类别化常被用作节省内存和提升性能的便捷工具。在筛选大型数据帧或系列后，许多类别可能不会出现在数据中。  
为帮助实现这一点，我们可以使用 `remove_unused_categories` 方法来删除未观察到的类别：

```python
cat_s3 = cat_s[cat_s.isin(['a', 'b'])]
cat_s3
# categories 'c' 和 'd' 依旧存在，占用内存
```

```python
cat_s3.cat.remove_unused_categories()
 # categories 'c' 和 'd' 被移除
```

### Creating dummy variables for modeling

当你使用统计或机器学习工具时，通常会将分类数据转换为 *dummy variables* ，也称为 *one-hot encoding*。  
这涉及创建一个 DataFrame，每个不同类别都有一列;这些列中，某一类别的出现次数为 `1`，其他类别为 `0`。

```python
cate_s = pd.Series(['a', 'b', 'c', 'd'] * 2, dtype='category')

pd.get_dummies(cate_s, dtype=float, prefix='category')
```
