## 使用函数或映射mapping转换数据

```python
data = pd.DataFrame({"food": ["bacon", "pulled pork", "bacon",
                              "pastrami", "corned beef", "bacon",
                              "pastrami", "honey ham", "nova lox"],
                     "ounces": [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

data
```

```python
data['food'].map(get_animal)
```

```python
data['food'].map(lambda x: meat_to_animal[x]) # lambda函数
```

## 离散化与分箱 Discretization and Binning

连续数据通常会被离散化或以其他方式分成“箱 bins”进行分析。

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
age_categories = pd.cut(ages, bins)

age_categories
```

```python
age_categories.codes
```

```python
age_categories.categories
```

```python
age_categories.value_counts()
```

在区间的字符串表示中，括号表示边是开的 （排他的），而方括号表示是闭的 （包含的）。

你可以通过传递 `right =False` 来改变哪一侧是闭合的：

```python
pd.cut(ages, bins, right=False)
```

```python
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]

pd.cut(ages, bins, labels=group_names)
```

如果你把整数个 bin 传递给 `pandas.cut` 而不是显式 bin 的边，它会根据数据中的最小值和最大值计算出等长 bin。

考虑一些均匀分布的数据被切成四分之一的情况：

```python
data = rng.uniform(size=20)
data_cate = pd.cut(data, 4, precision=2) # 小数点显示精度，只影响显示，不影响计算
# 第一个bin的下界是数据的最小值，最后一个 bin 的上界是数据的最大值
```

```python
data_cate.value_counts()
```

`pandas.qcut` 基于样本分位数对数据进行分组处理。根据数据分布不同，使用 `pandas.cut` 通常不会让每个箱的数据点数相同。由于 `pandas.qcut` 使用样本分位数，你会得到大小大致相同的箱：

```python
data = rng.standard_normal(1000)
quartiles = pd.qcut(data, 4, precision=2)

quartiles
```

```python
quartiles.value_counts()
```

```python
  可视化分箱过程
   1 数据点分布:   1    2    3    4                      100   101   102
   2              |    |    |    |-------- 巨大间隔 -----|    |    |
   3 
   4 pd.cut (等宽): |-------箱1-------|-------箱2-------|-------箱3-------|
   5               宽度相等但点数不均   空箱              点数集中
   6 
   7 pd.qcut (等频): |-箱1-| |-箱2-| |-----------箱3------------|
   8                 ≈2个点  ≈2个点            ≈3个点
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]).value_counts()
```

## 指示/虚拟变量 Computing Indicator/Dummy Variables

  核心思想: <br>将分类变量（categorical variable）
  转换为二进制（0/1）数值矩阵，使分类数据能够被数学模型（如回归、机器学习）处理。

  错误做法（序数编码）：  
   1 # 将分类直接映射为1,2,3...会导致模型误解  
   `2 df['city'] = df['city'].map({'北京': 1, '上海': 2, '广州': 3})`  
   3 # 问题：模型会认为"广州 > 上海 > 北京"（数值大小关系），但这毫无意义！

  正确做法（虚拟变量）：  
   1 # 创建独立的二进制列  
   `2 city_dummies = pd.get_dummies(df['city'], prefix='city')`  
   3 # 结果：city_北京, city_上海, city_广州 三列，每列值为0或1  

```python
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                   'data1': range(6)})
df
```

```python
pd.get_dummies(df['key'])
```

```python
pd.get_dummies(df['key'], dtype = 'float')
```

在 DataFrame 的列名显示前缀 `prefix`

```python
dummies = pd.get_dummies(df['key'], prefix='key', dtype=float)
df_with_dummies = df[['data1']].join(dummies)

df_with_dummies
```

关于为什么 `df[['data1']].join(dummies)` 要使用 `[[‘data1’]]`  

`df[[]]` 用于选择多列，返回 DataFrame  
=> `df[['a', 'b']]` ✅  
=> `df['a'. 'b']` ❌  

```python
print(df['data1'])
print(df['data1'].shape) # shape 是 Series，没有 .join(), .merge(), .groupby(), .pivot_table()..语法
print('-'*10)
print(df[['data1',]])
print(df[['data1',]].shape)  # shape 是 dataframe
print(df['data1', 'key'])  # 报错
```
